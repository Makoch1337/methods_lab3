## Прямая задача с барьером Кэрролла

Исходная прямая задача:

$$
\min_{x \in \mathbb{R}^n} \;\; f(x) = -\sum_{j=1}^{m} \pi_j \log(p_j^{\!\top} x), \qquad
\mathbf{1}^{\!\top} x = 1,\quad x \ge 0,
$$
где $\pi_j>0$, $\sum\pi_j=1$, и предполагается $p_j^{\!\top} x > 0$ (обычно $p_{j,i} > 0$).

Барьер Кэрролла для неравенств $x_i \ge 0$ имеет вид
$$
B(x) = \sum_{i=1}^{n} \frac{1}{x_i}\qquad (x>0).
$$
Задача с барьером (параметр $\mu>0$):

$$
\boxed{
\min_{x \in \mathbb{R}^n} \; F_{\mu}(x) = f(x) + \mu \sum_{i=1}^{n} \frac{1}{x_i}
\quad\text{при}\quad \mathbf{1}^{\!\top} x = 1,\; x > 0.} \tag{1}
$$

**Область определения:** $\mathcal{D}_{\text{pr}} = \{x\in\mathbb{R}_{++}^n \mid \mathbf{1}^{\!\top}x = 1\}$ --- относительная внутренность симплекса.

### Выпуклость

- $f(x)$ выпукла на $\mathcal{D}_{\text{pr}}$ (композиция выпуклой невозрастающей $-\log$ и аффинной $p_j^{\!\top}x$).
- $1/x_i$ выпукла на $x_i>0$, следовательно $\sum 1/x_i$ выпукла.
- $F_\mu$ --- сумма выпуклых функций с положительным весом $\mu$, поэтому **$F_\mu$ выпукла на $\mathcal{D}_{\text{pr}}$**.

### Градиент и гессиан

Градиент $f(x)$:
$$
\nabla f(x) = -\sum_{j=1}^{m} \frac{\pi_j}{p_j^{\!\top} x}\, p_j.
$$

Гессиан $f(x)$:
$$
\nabla^2 f(x) = \sum_{j=1}^{m} \frac{\pi_j}{(p_j^{\!\top} x)^2}\, p_j p_j^{\!\top} \;\succeq\; 0.
$$

Для барьерного слагаемого $r(x)=\sum_i 1/x_i$:
$$
\nabla r(x) = \left(-\frac{1}{x_1^2}, \dots, -\frac{1}{x_n^2}\right)^{\!\top},\qquad
\nabla^2 r(x) = \operatorname{diag}\left(\frac{2}{x_1^3}, \dots, \frac{2}{x_n^3}\right) \succ 0.
$$

Следовательно,
$$
\nabla F_\mu(x) = \nabla f(x) - \mu \begin{pmatrix} x_1^{-2} \\ \vdots \\ x_n^{-2} \end{pmatrix},
$$
$$
\nabla^2 F_\mu(x) = \nabla^2 f(x) + 2\mu \operatorname{diag}\bigl(x_1^{-3}, \dots, x_n^{-3}\bigr) \;\succ\; 0 \quad \text{на } \mathcal{D}_{\text{pr}}.
$$

### Строгая и сильная выпуклость

- **Исходная функция $f$** может не быть строго выпуклой на аффинном подпространстве: $\nabla^2 f(x)$ вырождена, если $m < n$ или векторы $p_j$ не порождают $\mathbb{R}^n$.  
- **Барьер** добавляет строго положительно определённую матрицу $\operatorname{diag}(2\mu/x_i^3)$. Вдоль любого направления $v \neq 0$ (даже при $\mathbf{1}^{\!\top}v=0$) имеем
  $$
  v^{\!\top}\nabla^2 F_\mu(x) v \ge 2\mu\sum_{i=1}^{n} \frac{v_i^2}{x_i^3} > 0,
  $$
  поскольку все $x_i>0$. Таким образом, **$F_\mu(x)$ строго выпукла на $\mathcal{D}_{\text{pr}}$**.
- **Сильная выпуклость** требует равномерной положительной определённости гессиана. На всём $\mathcal{D}_{\text{pr}}$ она отсутствует, так как $x_i \to 0$ или $x_i \to \infty$ приводят к неограниченному росту или убыванию кривизны.  
  Если ограничиться компактным подмножеством $x \in [\varepsilon, 1-\varepsilon]^n$ ($\varepsilon>0$), то
  $$
  \nabla^2 F_\mu(x) \succeq 2\mu\, \varepsilon^{-3} I \quad \text{на подпространстве } \mathbf{1}^{\!\top}v=0,
  $$
  то есть **на компакте $F_\mu$ сильно выпукла**.

- **Влияние на оптимизацию:** строгая выпуклость гарантирует единственность решения. Добавка барьера устраняет возможную вырожденность гессиана исходной задачи, обеспечивая обратимость $\nabla^2 F_\mu$ на всём $\mathcal{D}_{\text{pr}}$. Это принципиально для **метода Ньютона**: матрица становится положительно определённой, и система шага всегда имеет единственное решение.

## Двойственная задача в форме минимизации


Исходная прямая задача имела вид минимизации при ограничениях $x \ge 0$, $\mathbf{1}^\top x = 1$. Лагранжева двойственная функция, полученная снятием ограничений $y_j = p_j^\top x$, $\mathbf{1}^\top x = 1$ и $x \ge 0$, после устранения прямых переменных привела к следующей задаче максимизации:

$$
\max_{\nu > 0,\;\mu}\; \sum_{j=1}^{m} \pi_j \log \nu_j \;-\; \mu
\quad\text{при}\quad P^\top \nu \;\le\; \mu \mathbf{1}.
$$

Здесь $\nu \in \mathbb{R}^m$ --- множители Лагранжа к ограничениям $y_j = p_j^\top x$, а $\mu \in \mathbb{R}$ --- множитель к равенству $\mathbf{1}^\top x = 1$. Условие $P^\top \nu \le \mu \mathbf{1}$ возникло из требования неотрицательности множителей к неравенствам $x_i \ge 0$.

Чтобы перейти к задаче минимизации, вводится переменная $t \in \mathbb{R}$, заменяющая $\mu$, причём для любой допустимой пары $(\nu,\mu)$ выполнено $\mu \ge \max_i (P^\top \nu)_i$. Чтобы максимизировать $-\mu$, берётся минимальное возможное $\mu$, т.е. $\mu = \max_i (P^\top \nu)_i$. Тогда задача максимизации эквивалентна

$$
\max_{\nu > 0}\; \sum_{j=1}^{m} \pi_j \log \nu_j \;-\; \max_i \bigl(P^\top \nu\bigr)_i .
$$

Умножая целевую функцию на $-1$ и вводя переменную $t$, ограничивающую сверху все компоненты вектора $P^\top \nu$, получаем двойственную задачу в форме минимизации

$$
\boxed{
\begin{aligned}
\min_{\nu \in \mathbb{R}^m,\; t \in \mathbb{R}} \quad & D(\nu,t) = -\sum_{j=1}^{m} \pi_j \log \nu_j + t \\
\text{при} \quad & \nu_j > 0,\quad j=1,\ldots,m,\\
& \sum_{j=1}^{m} \nu_j\, p_{j,i} \;\le\; t,\quad i=1,\ldots,n .
\end{aligned}} \tag{2}
$$

Таким образом, переменными являются вектор $\nu$ длины $m$ и скаляр $t$, итого $m+1$ переменных. Ограничения задачи:

- $m$ строгих неравенств $\nu_j > 0$;
- $n$ неравенств $t - \sum_{j} \nu_j p_{j,i} \ge 0$, которые удобно записать в векторной форме
  $$
  t \mathbf{1} - P^\top \nu \;\ge\; 0,
  $$
  где $P$ --- матрица размера $n \times m$, составленная из столбцов $p_j$; элемент $(i,j)$ этой матрицы есть $p_{j,i}$.

Задача не содержит ограничений-равенств, что упрощает построение барьерного метода (отсутствие блоков $A$, $b$ в KKT‑системе).

Барьер Кэрролла для двойственной задачи
----------------------------------------

Барьер Кэрролла (обратный барьер) для набора неравенств $g_k(\nu,t) \le 0$ определяется как

$$
B_{\text{dual}}(\nu,t) \;=\; -\sum_{k} \frac{1}{g_k(\nu,t)} .
$$

Выпишем все ограничения в стандартной форме $g \le 0$:

- $-\nu_j \le 0$ для $j=1,\ldots,m$,
- $\displaystyle\sum_{j=1}^{m} \nu_j p_{j,i} - t \;\le\; 0$ для $i=1,\ldots,n$.

Подставляя в формулу, находим барьер

$$
\boxed{
B_{\text{dual}}(\nu,t) \;=\; \sum_{j=1}^{m} \frac{1}{\nu_j} \;+\; \sum_{i=1}^{n} \frac{1}{t - \sum_{j=1}^{m} \nu_j p_{j,i}} .
}
$$

Он определён и строго положителен только на внутренности допустимой области, т.е. при $\nu_j > 0$ и $t - \sum_j \nu_j p_{j,i} > 0$ для всех $i$. При приближении к любой границе хотя бы одно из слагаемых стремится к $+\infty$, удерживая итерации внутри области.

Регуляризованная двойственная задача
--------------------------------------

Вводя параметр барьера $\mu > 0$, строим семейство вспомогательных задач

$$
\boxed{
\begin{aligned}
\min_{\nu,\,t} \quad & \Phi_{\mu}(\nu,t) \;=\; D(\nu,t) \;+\; \mu\, B_{\text{dual}}(\nu,t) \\
\text{при} \quad & \nu > 0,\quad t - \sum_{j=1}^{m} \nu_j p_{j,i} > 0,\;\; i=1,\ldots,n .
\end{aligned}} \tag{3}
$$

Здесь $\Phi_{\mu}$ --- строго выпуклая функция на выпуклой области

$$
\mathcal{D}_{\text{dual}} \;=\; \bigl\{ (\nu,t) \in \mathbb{R}^m \times \mathbb{R} \;\big|\; \nu>0,\; t > \max_i (P^\top \nu)_i \bigr\},
$$

поскольку $D(\nu,t)$ выпукла, а каждое слагаемое барьера --- выпуклая функция (композиция выпуклой убывающей $1/s$ и аффинного отображения). Сумма выпуклых функций с положительным коэффициентом $\mu$ сохраняет выпуклость.

С ростом $\mu$ влияние барьера ослабевает, и при $\mu \to \infty$ решение задачи (3) сходится к решению исходной двойственной задачи (2). На практике $\mu$ увеличивают на каждой внешней итерации барьерного метода, решая подзадачу (3) с заданным $\mu$ с высокой точностью (обычно методом Ньютона).

### Выпуклость $\Phi_{\mu}$

- $D(\nu,t)$ выпукла: $-\log\nu_j$ выпукла, $t$ линейно.
- $1/\nu_j$ выпукла на $\nu_j>0$.
- Функция $\eta_i(\nu,t) = t - \sum_j \nu_j p_{j,i}$ аффинна; $h(s) = 1/s$ выпукла и убывает при $s>0$, её композиция с аффинным отображением выпукла. Следовательно, каждое слагаемое $1/\bigl(t - \sum_j \nu_j p_{j,i}\bigr)$ выпукло.
- Сумма выпуклых функций с положительными весами сохраняет выпуклость.  
Итак, **$\Phi_{\mu}$ выпукла на $\mathcal{D}_{\text{dual}}$**.

### Градиент и гессиан

Введём обозначения:
$$
s_i = t - \sum_{j=1}^{m} \nu_j p_{j,i} \quad (>0).
$$
Тогда
$$
\Phi_{\mu} = \underbrace{-\sum_{j} \pi_j \log \nu_j + t}_{D} \;+\; \mu\sum_{j} \frac{1}{\nu_j} \;+\; \mu\sum_{i} \frac{1}{s_i}.
$$

**Градиент:**  
По $\nu_k$ ($k=1,\dots,m$):
$$
\frac{\partial D}{\partial \nu_k} = -\frac{\pi_k}{\nu_k}, \qquad
\frac{\partial}{\partial \nu_k}\!\left(\mu\sum_{j}\frac{1}{\nu_j}\right) = -\mu\frac{1}{\nu_k^2},
$$
$$
\frac{\partial}{\partial \nu_k}\!\left(\mu\sum_{i}\frac{1}{s_i}\right) = \mu\sum_{i} \frac{1}{s_i^2} \cdot p_{k,i}
\quad (\text{так как } \partial s_i/\partial \nu_k = -p_{k,i}).
$$
По $t$:
$$
\frac{\partial D}{\partial t} = 1,\qquad
\frac{\partial}{\partial t}\!\left(\mu\sum_i \frac{1}{s_i}\right) = -\mu\sum_i \frac{1}{s_i^2}.
$$
Барьерные члены от $\nu_j$ не зависят от $t$.

Итак,
$$
\frac{\partial \Phi_{\mu}}{\partial \nu_k} = -\frac{\pi_k}{\nu_k} - \frac{\mu}{\nu_k^2} + \mu\sum_{i=1}^{n} \frac{p_{k,i}}{s_i^2},
\qquad
\frac{\partial \Phi_{\mu}}{\partial t} = 1 - \mu\sum_{i=1}^{n} \frac{1}{s_i^2}.
$$

**Гессиан:**  
Рассмотрим блоки матрицы вторых производных. Размерность $m+1$, индексация: сначала $\nu$, затем $t$.

- Блок $\nabla_{\nu}^2$ ($m\times m$):
  От $D$: $\operatorname{diag}(\pi_k/\nu_k^2)$.
  От барьера $\mu\sum 1/\nu_j$: $\operatorname{diag}(2\mu/\nu_k^3)$.
  От барьера $\mu\sum 1/s_i$: для $k,l$
  $$
  \frac{\partial^2}{\partial \nu_k \partial \nu_l} \frac{1}{s_i} =
  \frac{\partial}{\partial \nu_l}\!\left( \frac{p_{k,i}}{s_i^2} \right) =
  2\frac{p_{k,i} p_{l,i}}{s_i^3}.
  $$
  Суммируя по $i$, получаем матрицу $H_{\nu} = 2\mu \sum_{i=1}^n \frac{1}{s_i^3} \, \mathbf{p}_i \mathbf{p}_i^{\!\top}$, где $\mathbf{p}_i$ --- $i$-я строка матрицы $P$ размером $n\times m$ (или, точнее, вектор $\bigl(p_{1,i}, \dots, p_{m,i}\bigr)^{\!\top}$).

  Таким образом,
  $$
  \nabla_{\nu}^2 \Phi_{\mu} = \operatorname{diag}\!\left(\frac{\pi_k}{\nu_k^2} + \frac{2\mu}{\nu_k^3}\right)_{k=1}^m \;+\; 2\mu \sum_{i=1}^{n} \frac{1}{s_i^3}\, \mathbf{p}_i \mathbf{p}_i^{\!\top}.
  $$

- Смешанный блок $\nabla_{\nu t}^2$ (вектор длины $m$):
  $$
  \frac{\partial^2 \Phi_{\mu}}{\partial \nu_k \partial t} = \frac{\partial}{\partial t}\!\left( \mu\sum_i \frac{p_{k,i}}{s_i^2} \right) = -2\mu\sum_i \frac{p_{k,i}}{s_i^3}.
  $$
  Здесь использовано $\partial s_i/\partial t = 1$ и производная $s_i^{-2}$ по $t$.

- Вторая производная по $t$ (скаляр):
  $$
  \frac{\partial^2 \Phi_{\mu}}{\partial t^2} = \frac{\partial}{\partial t}\!\left(1 - \mu\sum_i \frac{1}{s_i^2}\right) = 2\mu\sum_{i=1}^{n} \frac{1}{s_i^3}.
  $$

Полный гессиан $\nabla^2 \Phi_{\mu} \in \mathbb{R}^{(m+1)\times(m+1)}$ записывается как
$$
\nabla^2 \Phi_{\mu} =
\begin{pmatrix}
\nabla_{\nu}^2 \Phi_{\mu} & \mathbf{g} \\
\mathbf{g}^{\!\top} & 2\mu\sum_i s_i^{-3}
\end{pmatrix},
$$
где $\mathbf{g} = -2\mu \bigl( \sum_i s_i^{-3} p_{1,i},\; \dots,\; \sum_i s_i^{-3} p_{m,i} \bigr)^{\!\top}$.

### Свойства выпуклости гессиана

- Матрица $\nabla_{\nu}^2 \Phi_{\mu}$ всегда положительно определена: слагаемые $\operatorname{diag}(\pi_k/\nu_k^2 + 2\mu/\nu_k^3)$ дают строго положительную диагональ, а $2\mu\sum s_i^{-3} \mathbf{p}_i \mathbf{p}_i^{\!\top} \succeq 0$.
- Полный гессиан может быть записан как сумма диагональной положительно определённой матрицы (по $\nu$) и нескольких неотрицательно определённых матриц ранга 1, дополненных строкой/столбцом. В целом, $\nabla^2 \Phi_{\mu} \succ 0$ на всей $\mathcal{D}_{\text{dual}}$, так как для любого вектора $(v^\top, w)^\top \neq 0$ квадратичная форма строго положительна (положительность по $\nu$ гарантирует её, если $v\neq 0$; если же $v=0$, то $w\neq 0$, и вклад даёт $2\mu\sum s_i^{-3} w^2 > 0$). Таким образом, **$\Phi_{\mu}$ строго выпукла**.
- **Сильная выпуклость** отсутствует глобально: при $\nu_j \to \infty$ или $s_i \to \infty$ минимальное собственное число стремится к нулю. На любом компакте, отделённом от границ, функция сильно выпукла.

### Влияние барьера и применимость методов второго порядка

- **Исходная двойственная функция $D(\nu,t)$**, как правило, не является строго выпуклой: по переменной $t$ гессиан равен нулю (нулевая строка/столбец), а $\nabla_{\nu}^2 D = \operatorname{diag}(\pi_j/\nu_j^2)$ --- положительно определена только на $\nu$, но всё вместе даёт вырожденную матрицу (ранг $m$). Это препятствует прямому применению метода Ньютона.
- **Барьер Кэрролла** добавляет положительную кривизну по $t$ (слагаемое $2\mu\sum s_i^{-3}$) и смешанные производные, делая полный гессиан **обратимым и положительно определённым**. Теперь метод Ньютона корректен: шаг $\Delta = -(\nabla^2 \Phi_{\mu})^{-1} \nabla \Phi_{\mu}$ определён.
- Добавка $\mu\sum 1/\nu_j$ усиливает диагональное преобладание в блоке $\nu$, улучшая обусловленность матрицы (число обусловленности падает при больших $\mu$).
- С практической точки зрения при выборе между прямой и двойственной задачами с барьером:
  * Размерность прямой: $n$ (число активов).
  * Размерность двойственной: $m+1$ (число сценариев плюс один).
  * Барьер в прямой задаче добавляет всего $n$ слагаемых (по одному на актив), в двойственной --- $m+n$ слагаемых, что дополнительно увеличивает стоимость вычисления гессиана.

```python
import numpy as np
import cvxpy as cp
import pandas as pd
import time
import matplotlib.pyplot as plt

from abc import ABC, abstractmethod
from typing import Dict, Any, List, Tuple

class BarrierOptimizationProblem(ABC):
    @abstractmethod
    def f_obj(self, x: np.ndarray) -> float:
        pass

    @abstractmethod
    def grad_obj(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def hess_obj(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def f_barrier(self, x: np.ndarray) -> float:
        pass

    @abstractmethod
    def grad_barrier(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def hess_barrier(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def constraints(self) -> Tuple[np.ndarray, np.ndarray]:
        pass

    @abstractmethod
    def generate_start_points(self, num_points: int) -> List[np.ndarray]:
        pass

class Optimizer(ABC):
    @abstractmethod
    def solve(self, problem: BarrierOptimizationProblem, x0: np.ndarray, **kwargs) -> Dict[str, Any]:
        pass

    def test(self, problem: BarrierOptimizationProblem, x0: np.ndarray, expected_val: float, tol: float = 1e-2, **kwargs) -> bool:
        try:
            res = self.solve(problem, x0, **kwargs)
            is_passed = abs(res['opt_val'] - expected_val) <= tol
            if not is_passed:
                print(f"[{self.__str__()}] Test Failed! Expected: {expected_val:.4f}, Got: {res['opt_val']:.4f}")
            return is_passed
        except Exception as e:
            print(f"[{self.__str__()}] Test Failed with exception: {e}")
            return False

    @abstractmethod
    def __str__(self) -> str:
        pass

class DataGenerator(ABC):
    @abstractmethod
    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        pass

    @abstractmethod
    def __str__(self) -> str:
        pass
```

```python
class LogOptimalPrimalWithBarrier(BarrierOptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray):
        self.P = P
        self.pi = pi
        self.n, self.m = P.shape

    def f_obj(self, x: np.ndarray) -> float:
        y = self.P.T @ x
        if np.any(y <= 1e-12):
            return np.inf
        return -np.sum(self.pi * np.log(y))

    def grad_obj(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        return -self.P @ (self.pi / denom)

    def hess_obj(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        tmp = self.pi / (denom ** 2)
        return self.P @ np.diag(tmp) @ self.P.T

    def f_barrier(self, x: np.ndarray) -> float:
        if np.any(x <= 1e-12):
            return np.inf
        return np.sum(1.0 / x)

    def grad_barrier(self, x: np.ndarray) -> np.ndarray:
        return -1.0 / (x**2 + 1e-16)

    def hess_barrier(self, x: np.ndarray) -> np.ndarray:
        return np.diag(2.0 / (x**3 + 1e-16))

    def constraints(self) -> Tuple[np.ndarray, np.ndarray]:
        A = np.ones((1, self.n))
        b = np.array([1.0])
        return A, b

    def generate_start_points(self, num_points: int) -> List[np.ndarray]:
        points = []
        points.append(np.ones(self.n) / self.n)

        for _ in range(num_points - 1):0.028809	11	459	88

            x = np.random.rand(self.n) + 0.1
            x /= np.sum(x)
            points.append(x)
        return points

class LogOptimalDualWithBarrier(BarrierOptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray):
        self.P = P
        self.pi = pi
        self.n, self.m = P.shape
        self.n_var = self.m + 1

    def _unpack(self, z: np.ndarray):
        return z[:self.m], z[self.m]

    def f_obj(self, z: np.ndarray) -> float:
        w, nu = self._unpack(z)
        if np.any(w <= 1e-12):
            return -np.inf
        return np.sum(self.pi * np.log(w)) - nu

    def grad_obj(self, z: np.ndarray) -> np.ndarray:
        w, nu = self._unpack(z)
        grad_w = self.pi / (w + 1e-16)
        grad_nu = -1.0
        return np.concatenate([grad_w, [grad_nu]])

    def hess_obj(self, z: np.ndarray) -> np.ndarray:
        w, nu = self._unpack(z)
        H = np.zeros((self.n_var, self.n_var))
        H[:self.m, :self.m] = -np.diag(self.pi / (w**2 + 1e-16))
        return H

    def f_barrier(self, z: np.ndarray) -> float:
        w, nu = self._unpack(z)
        margin = nu - self.P @ w
        if np.any(w <= 1e-12) or np.any(margin <= 1e-12):
            return np.inf
        return np.sum(1.0 / margin) + np.sum(1.0 / w)

    def grad_barrier(self, z: np.ndarray) -> np.ndarray:
        w, nu = self._unpack(z)
        margin = nu - self.P @ w + 1e-16
        grad_nu = -np.sum(1.0 / (margin**2))
        grad_w = self.P.T @ (1.0 / (margin**2)) - 1.0 / (w**2 + 1e-16)
        return np.concatenate([grad_w, [grad_nu]])

    def hess_barrier(self, z: np.ndarray) -> np.ndarray:
        w, nu = self._unpack(z)
        margin = nu - self.P @ w + 1e-16
        inv_margin_cube = 2.0 / (margin**3)

        H_nu_nu = np.sum(inv_margin_cube)
        H_w_nu = -self.P.T @ inv_margin_cube
        H_w_w = self.P.T @ np.diag(inv_margin_cube) @ self.P + np.diag(2.0 / (w**3 + 1e-16))

        H = np.zeros((self.n_var, self.n_var))
        H[:self.m, :self.m] = H_w_w
        H[:self.m, self.m] = H_w_nu
        H[self.m, :self.m] = H_w_nu
        H[self.m, self.m] = H_nu_nu
        return H

    def constraints(self) -> Tuple[np.ndarray, np.ndarray]:
        return np.zeros((0, self.n_var)), np.zeros(0)

    def generate_start_points(self, num_points: int) -> List[np.ndarray]:
        points = []
        for i in range(num_points):
            w = np.ones(self.m) * np.mean(self.pi)
            margin = np.random.rand() * 5 + 1.0
            initial_nu = np.mean(self.P @ w) + (np.max(self.P @ w) - np.mean(self.P @ w)) + margin
            points.append(np.concatenate([w, [initial_nu]]))
        return points
```

```python
class NormalDataGenerator(DataGenerator):
    def __init__(self, scale: float = 0.1, shift: float = 1.0):
        self.scale = scale
        self.shift = shift

    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        m = n * 2
        P = np.random.randn(n, m) * self.scale + self.shift
        P += np.eye(n, m) * n
        pi = np.random.rand(m)
        pi /= pi.sum()
        return P, pi

    def __str__(self) -> str:
        return f"Normal(scale={self.scale}, shift={self.shift})"

class UniformDataGenerator(DataGenerator):
    def __init__(self, low: float = 0.9, high: float = 1.1):
        self.low = low
        self.high = high

    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        m = n * 2
        P = np.random.uniform(self.low, self.high, size=(n, m))
        P += np.eye(n, m) * n
        pi = np.random.rand(m)
        pi /= pi.sum()
        return P, pi

    def __str__(self) -> str:
        return f"Uniform(low={self.low}, high={self.high})"
```

```python
class CVXPYOptimizer(Optimizer):
    def __init__(self, solver: str = 'ECOS'):
        self.solver = solver

    def solve(self, problem: BarrierOptimizationProblem, x0: np.ndarray = None, **kwargs) -> Dict[str, Any]:
        start_time = time.time()

        if isinstance(problem, LogOptimalPrimalWithBarrier):
            x = cp.Variable(problem.n)
            objective = cp.Minimize(-cp.sum(cp.multiply(problem.pi, cp.log(problem.P.T @ x))))
            constraints = [cp.sum(x) == 1, x >= 0]
            cvx_prob = cp.Problem(objective, constraints)

        elif isinstance(problem, LogOptimalDualWithBarrier):
            w = cp.Variable(problem.m)
            nu = cp.Variable()
            objective = cp.Maximize(cp.sum(cp.multiply(problem.pi, cp.log(w))) - nu)
            constraints = [problem.P @ w - nu <= 0, w >= 0]
            cvx_prob = cp.Problem(objective, constraints)

        cvx_prob.solve(solver=self.solver)
        end_time = time.time()

        return {
            'opt_val': cvx_prob.value,
            'opt_var': x.value if isinstance(problem, LogOptimalPrimalWithBarrier) else np.concatenate([w.value, [nu.value]]),
            'time': end_time - start_time,
            'iters': cvx_prob.solver_stats.num_iters if cvx_prob.solver_stats else 0,
            'status': cvx_prob.status,
            'history': [cvx_prob.value]
        }

    def __str__(self) -> str:
        return f"CVXPY({self.solver})"
```

```python
class NewtonBarrierOptimizer(Optimizer):
    def __init__(self,
                 eps_inner: float = 1e-8,
                 alpha: float = 0.25,
                 beta: float = 0.5,
                 max_iters_inner: int = 200,
                 max_iters_outer: int = 200):
        self.eps_inner = eps_inner
        self.alpha = alpha
        self.beta = beta
        self.max_iters_inner = max_iters_inner
        self.max_iters_outer = max_iters_outer

    def solve(self, problem: BarrierOptimizationProblem, x0: np.ndarray,
              t0: float = 1.0, mu: float = 2.0, eps_outer: float = 1e-6, **kwargs) -> Dict[str, Any]:

        start_time = time.time()
        x = x0.copy()
        t = t0
        history = []
        total_inner_iters = 0
        prev_f = float('inf')

        sign = 1.0 if isinstance(problem, LogOptimalPrimalWithBarrier) else -1.0

        A, b = problem.constraints()
        has_constraints = A.shape[0] > 0

        for outer_iter in range(self.max_iters_outer):
            history.append(problem.f_obj(x))

            for inner_iter in range(self.max_iters_inner):
                g = t * sign * problem.grad_obj(x) + problem.grad_barrier(x)
                H = t * sign * problem.hess_obj(x) + problem.hess_barrier(x)

                if has_constraints:
                    kkt_matrix = np.block([
                        [H, A.T],
                        [A, np.zeros((A.shape[0], A.shape[0]))]
                    ])
                    rhs = np.concatenate([-g, np.zeros(A.shape[0])])
                    try:
                        sol = np.linalg.solve(kkt_matrix, rhs)
                        dx = sol[:len(x)]
                    except:
                        dx = np.linalg.lstsq(kkt_matrix, rhs, rcond=None)[0][:len(x)]
                else:
                    try:
                        dx = np.linalg.solve(H, -g)
                    except:
                        dx = np.linalg.lstsq(H, -g, rcond=None)[0]

                lambda_sq = dx @ H @ dx
                if lambda_sq / 2.0 <= self.eps_inner:
                    break

                step_size = 1.0
                f_current = t * sign * problem.f_obj(x) + problem.f_barrier(x)

                while True:
                    x_next = x + step_size * dx

                    f_bar_next = problem.f_barrier(x_next)

                    if not np.isfinite(f_bar_next):
                        step_size *= self.beta
                        continue

                    f_obj_next = problem.f_obj(x_next)

                    if not np.isfinite(f_obj_next):
                        step_size *= self.beta
                        continue

                    f_next_total = t * sign * f_obj_next + f_bar_next

                    if f_next_total <= f_current + self.alpha * step_size * (g @ dx):
                        break

                    step_size *= self.beta
                    if step_size < 1e-15:
                        break

                x = x + step_size * dx
                total_inner_iters += 1

            #current_f = problem.f_obj(x)
            #if abs(current_f - prev_f) < eps_outer:
            #    break
            #prev_f = current_f

            if problem.f_barrier(x) / t < eps_outer:
                break

            t *= mu

        end_time = time.time()

        return {
            'opt_val': problem.f_obj(x),
            'opt_var': x,
            'time': end_time - start_time,
            'iters': outer_iter + 1,
            'total_inner_iters': total_inner_iters,
            'status': 'converged',
            'history': history
        }

    def __str__(self) -> str:
        return "Barrier (Newton)"
```

```python
class NewtonBarrierOptimizerByGap(Optimizer):
    def __init__(self,
                 eps_inner: float = 1e-8,
                 alpha: float = 0.25,
                 beta: float = 0.5,
                 max_iters_inner: int = 200,
                 max_iters_outer: int = 200):
        self.eps_inner = eps_inner
        self.alpha = alpha
        self.beta = beta
        self.max_iters_inner = max_iters_inner
        self.max_iters_outer = max_iters_outer

    def solve(self, problem: BarrierOptimizationProblem, x0: np.ndarray,
              t0: float = 1.0, mu: float = 2.0, eps_outer: float = 1e-6, **kwargs) -> Dict[str, Any]:

        start_time = time.time()
        x = x0.copy()
        t = t0
        history = []
        total_inner_iters = 0
        prev_f = float('inf')

        sign = 1.0 if isinstance(problem, LogOptimalPrimalWithBarrier) else -1.0

        A, b = problem.constraints()
        has_constraints = A.shape[0] > 0

        for outer_iter in range(self.max_iters_outer):
            history.append(problem.f_obj(x))

            for inner_iter in range(self.max_iters_inner):
                g = t * sign * problem.grad_obj(x) + problem.grad_barrier(x)
                H = t * sign * problem.hess_obj(x) + problem.hess_barrier(x)

                if has_constraints:
                    kkt_matrix = np.block([
                        [H, A.T],
                        [A, np.zeros((A.shape[0], A.shape[0]))]
                    ])
                    rhs = np.concatenate([-g, np.zeros(A.shape[0])])
                    try:
                        sol = np.linalg.solve(kkt_matrix, rhs)
                        dx = sol[:len(x)]
                    except:
                        dx = np.linalg.lstsq(kkt_matrix, rhs, rcond=None)[0][:len(x)]
                else:
                    try:
                        dx = np.linalg.solve(H, -g)
                    except:
                        dx = np.linalg.lstsq(H, -g, rcond=None)[0]

                lambda_sq = dx @ H @ dx
                if lambda_sq / 2.0 <= self.eps_inner:
                    break

                step_size = 1.0
                f_current = t * sign * problem.f_obj(x) + problem.f_barrier(x)

                while True:
                    x_next = x + step_size * dx

                    f_bar_next = problem.f_barrier(x_next)

                    if not np.isfinite(f_bar_next):
                        step_size *= self.beta
                        continue

                    f_obj_next = problem.f_obj(x_next)

                    if not np.isfinite(f_obj_next):
                        step_size *= self.beta
                        continue

                    f_next_total = t * sign * f_obj_next + f_bar_next

                    if f_next_total <= f_current + self.alpha * step_size * (g @ dx):
                        break

                    step_size *= self.beta
                    if step_size < 1e-15:
                        break

                x = x + step_size * dx
                total_inner_iters += 1

            if isinstance(problem, LogOptimalPrimalWithBarrier):
                y = problem.P.T @ x
                nu = problem.pi / y
                mu_val = np.max(problem.P @ nu)
                gap = mu_val - 1.0
            else:
                w, nu_val = x[:problem.m], x[problem.m]
                s = nu_val - problem.P @ w
                x_primal = 1.0 / (t * s**2)
                x_primal = x_primal / np.sum(x_primal)
                f_primal = -np.sum(problem.pi * np.log(problem.P.T @ x_primal + 1e-16))
                C = np.sum(problem.pi * (1.0 - np.log(problem.pi)))
                g_dual = np.sum(problem.pi * np.log(w)) - nu_val + C
                gap = f_primal - g_dual

            if gap < eps_outer:
                break

            t *= mu

        end_time = time.time()

        return {
            'opt_val': problem.f_obj(x),
            'opt_var': x,
            'time': end_time - start_time,
            'iters': outer_iter + 1,
            'total_inner_iters': total_inner_iters,
            'status': 'converged',
            'history': history
        }

    def __str__(self) -> str:
        return "Barrier (Newton)"
```

```python
class QuasiNewtonBarrierOptimizer(Optimizer):

    def __init__(self,
                 eps_inner: float = 1e-8,
                 alpha: float = 0.25,
                 beta: float = 0.5,
                 max_iters_inner: int = 200,
                 max_iters_outer: int = 200):
        self.eps_inner = eps_inner
        self.alpha = alpha
        self.beta = beta
        self.max_iters_inner = max_iters_inner
        self.max_iters_outer = max_iters_outer

    @staticmethod
    def _projector(A: np.ndarray) -> np.ndarray:
        AAT_inv = np.linalg.inv(A @ A.T)
        return np.eye(A.shape[1]) - A.T @ AAT_inv @ A

    @staticmethod
    def _bfgs_update(B: np.ndarray, s: np.ndarray, y: np.ndarray) -> np.ndarray:
        sy = s @ y
        if sy <= 1e-10:
            return B
        rho = 1.0 / sy
        I = np.eye(len(s))
        V = I - rho * np.outer(s, y)
        return V @ B @ V.T + rho * np.outer(s, s)

    def solve(self, problem: BarrierOptimizationProblem, x0: np.ndarray,
              t0: float = 1.0, mu: float = 2.0, eps_outer: float = 1e-6, **kwargs) -> Dict[str, Any]:
        start_time = time.time()
        x = x0.copy()
        t = t0
        history = []
        total_inner_iters = 0

        sign = 1.0 if isinstance(problem, LogOptimalPrimalWithBarrier) else -1.0

        A, b = problem.constraints()
        has_constraints = A.shape[0] > 0
        P_null = self._projector(A) if has_constraints else None
        n = len(x)

        B = None

        for outer_iter in range(self.max_iters_outer):
            history.append(problem.f_obj(x))

            # t*sign*f_obj растёт в mu раз, поэтому обратный гессиан ~ B/mu.
            if B is not None and outer_iter > 0:
                B = B / mu

            g_full = t * sign * problem.grad_obj(x) + problem.grad_barrier(x)
            g = P_null @ g_full if has_constraints else g_full
            phi_prev = t * sign * problem.f_obj(x) + problem.f_barrier(x)

            for inner_iter in range(self.max_iters_inner):
                if B is None:
                    dx = -g
                else:
                    dx = -B @ g
                if has_constraints:
                    dx = P_null @ dx

                lambda_sq = -g @ dx
                if lambda_sq / 2.0 <= self.eps_inner:
                    break

                step_size = 1.0
                f_current = t * sign * problem.f_obj(x) + problem.f_barrier(x)
                directional = g @ dx

                while True:
                    x_next = x + step_size * dx
                    f_bar_next = problem.f_barrier(x_next)
                    if not np.isfinite(f_bar_next):
                        step_size *= self.beta
                        if step_size < 1e-15:
                            break
                        continue

                    f_obj_next = problem.f_obj(x_next)
                    if not np.isfinite(f_obj_next):
                        step_size *= self.beta
                        if step_size < 1e-15:
                            break
                        continue

                    f_next_total = t * sign * f_obj_next + f_bar_next
                    if f_next_total <= f_current + self.alpha * step_size * directional:
                        break

                    step_size *= self.beta
                    if step_size < 1e-15:
                        break

                if step_size < 1e-15:
                    break

                x_new = x + step_size * dx

                g_full_new = t * sign * problem.grad_obj(x_new) + problem.grad_barrier(x_new)
                g_new = P_null @ g_full_new if has_constraints else g_full_new

                s_vec = x_new - x
                y_vec = g_new - g

                if B is None:
                    sy0 = s_vec @ y_vec
                    yy0 = y_vec @ y_vec
                    if sy0 > 1e-12 and yy0 > 1e-12:
                        B = (sy0 / yy0) * np.eye(n)
                    else:
                        B = np.eye(n)

                B = self._bfgs_update(B, s_vec, y_vec)

                x = x_new
                g = g_new
                total_inner_iters += 1

                phi_new = t * sign * problem.f_obj(x) + problem.f_barrier(x)
                if abs(phi_new - phi_prev) < 1e-10 * max(1.0, abs(phi_new)):
                    break
                phi_prev = phi_new

            if problem.f_barrier(x) / t < eps_outer:
                break

            t *= mu

        end_time = time.time()

        return {
            'opt_val': problem.f_obj(x),
            'opt_var': x,
            'time': end_time - start_time,
            'iters': outer_iter + 1,
            'total_inner_iters': total_inner_iters,
            'status': 'converged',
            'history': history
        }

    def __str__(self) -> str:
        return "Barrier (Quasi-Newton)"
```

```python
class ExperimentRunner:
    def __init__(self,
                 n_list: List[int],
                 cases_per_n: int,
                 start_points_count: int,
                 ground_truth_solver: Optimizer,
                 optimizers: List[Optimizer],
                 generators: List[DataGenerator],
                 t0_list: List[float] = [1.0],
                 mu_list: List[float] = [15.0]):
        self.n_list = n_list
        self.cases_per_n = cases_per_n
        self.start_points_count = start_points_count
        self.ground_truth_solver = ground_truth_solver
        self.optimizers = optimizers
        self.generators = generators
        self.t0_list = t0_list
        self.mu_list = mu_list

    def run(self) -> pd.DataFrame:
        rows = []

        for n in self.n_list:
            for gen in self.generators:
                for case in range(self.cases_per_n):
                    P, pi = gen.generate(n)
                    primal = LogOptimalPrimalWithBarrier(P, pi)
                    dual = LogOptimalDualWithBarrier(P, pi)

                    p_start_points = primal.generate_start_points(self.start_points_count)
                    d_start_points = dual.generate_start_points(self.start_points_count)

                    gt_primal = self.ground_truth_solver.solve(primal, x0=p_start_points[0])
                    gt_dual = self.ground_truth_solver.solve(dual, x0=d_start_points[0])

                    for t0 in self.t0_list:
                        for mu in self.mu_list:
                            for sp_idx in range(self.start_points_count):
                                x0_p = p_start_points[sp_idx]
                                z0_d = d_start_points[sp_idx]
                                for opt in self.optimizers:

                                    if not opt.test(primal, x0_p, gt_primal['opt_val'], t0=t0, mu=mu) or not opt.test(dual, z0_d, gt_dual['opt_val'], t0=t0, mu=mu):
                                        pass

                                    try:
                                        res_primal = opt.solve(primal, x0=x0_p, t0=t0, mu=mu)
                                        res_dual = opt.solve(dual, x0=z0_d, t0=t0, mu=mu)

                                        rows.append({
                                            'n': n, 'case': case, 'generator': str(gen), 'solver': str(opt),
                                            'start_point': sp_idx, 't0': t0, 'mu': mu,
                                            'gt_p_opt': gt_primal['opt_val'], 'gt_d_opt': gt_dual['opt_val'],
                                            'p_opt': res_primal['opt_val'], 'd_opt': res_dual['opt_val'],
                                            'time_primal': res_primal['time'], 'iters_primal': res_primal['iters'],
                                            'time_dual': res_dual['time'], 'iters_dual': res_dual['iters'],
                                            'inner_iters_primal': res_primal['total_inner_iters'],  
                                            'inner_iters_dual': res_dual['total_inner_iters'], 
                                        })
                                    except NotImplementedError:
                                        pass

        # TODO: добавить вывод графиков

        return pd.DataFrame(rows)
```

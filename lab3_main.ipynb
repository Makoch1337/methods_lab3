### Прямая задача (дополнительное ограничение $\|x\|^2 \leq R^2$)

$$
\begin{aligned}
\min_{x\in\mathbb{R}^n}\quad & -\sum_{j=1}^m \pi_j \ln(p_j^\top x) \\
\text{s.t.}\quad & \mathbf{1}^\top x = 1, \\
& p_j^\top x \ge \varepsilon > 0,\quad j=1,\dots,m,\\
& \|x\|_2 \le R.
\end{aligned}
\tag{P}
$$

**Выпуклость**  
- $-\ln t$ выпукла на $(0,\infty)$;  
- композиция $-\ln(p_j^\top x)$ с аффинной $p_j^\top x$ выпукла на $\{x\mid p_j^\top x>0\}$;  
- на $\mathcal{F}$ имеем $p_j^\top x\ge\varepsilon>0$, поэтому каждое слагаемое $-\pi_j\ln(p_j^\top x)$ выпукло;  
- сумма выпуклых функций с неотрицательными весами $\pi_j$ выпукла $\Rightarrow$ $f$ выпукла;  
- $\mathcal{F}$ -- пересечение аффинного подпространства, полупространств и шара -- замкнуто, выпукло и компактно.

**Условие Слейтера**  
Предположим, что существует портфель $\bar{x}$ с $\mathbf{1}^\top\bar{x}=1$ и $p_j^\top\bar{x}>0$ для всех $j$.
Выбрав $\varepsilon\in(0,\;\min_j p_j^\top\bar{x})$ и $R>\|\bar{x}\|_2$, получим строго допустимую точку:
$$
\mathbf{1}^\top\bar{x}=1,\quad p_j^\top\bar{x}>\varepsilon\;\forall j,\quad \|\bar{x}\|_2<R,
$$
то есть условие Слейтера выполнено.

**Сильная двойственность**  
Задача выпукла, $\mathcal{F}$ непусто и компактно, условие Слейтера выполнено $\Longrightarrow$ сильная двойственность: оптимальные значения прямой и двойственной задач совпадают.

---

### Введение вспомогательных переменных

Введём $s_j = p_j^\top x$. Задача (P) эквивалентна

$$
\begin{aligned}
\min_{x,\,s}\quad & -\sum_{j=1}^m \pi_j \ln s_j \\
\text{s.t.}\quad & s_j = p_j^\top x, \\
& \mathbf{1}^\top x = 1, \\
& s_j \ge \varepsilon, \\
& \|x\|_2 \le R.
\end{aligned}
\tag{1}
$$

Целевая функция зависит только от $s$, ограничения связывают $x$ и $s$.

---

### Лагранжиан

Двойственные переменные:
- $\nu\in\mathbb{R}$ для $\mathbf{1}^\top x = 1$,
- $\alpha_j\in\mathbb{R}$ для $s_j = p_j^\top x$,
- $\lambda_j\ge 0$ для $s_j \ge \varepsilon$,
- $\mu\ge 0$ для $\|x\|_2^2 \le R^2$.

Лагранжиан (ограничения‑неравенства приведены к виду $g(x)\le 0$):

$$
\begin{aligned}
L(x,s,\nu,\alpha,\lambda,\mu) = &-\sum_{j}\pi_j\ln s_j \\
&+ \nu(\mathbf{1}^\top x - 1) \\
&+ \sum_{j}\alpha_j(p_j^\top x - s_j) \\
&+ \sum_{j}\lambda_j(\varepsilon - s_j) \\
&+ \mu\bigl(\|x\|_2^2 - R^2\bigr).
\end{aligned}
$$

Перегруппируем слагаемые, выделяя части, зависящие только от $x$ и только от $s$:

$$
L = \underbrace{\bigl[\mu\|x\|_2^2 + (P\alpha + \nu\mathbf{1})^\top x\bigr]}_{\text{часть }(x)} \;+\; \underbrace{\sum_{j}\bigl[-\pi_j\ln s_j - (\alpha_j+\lambda_j)s_j + \lambda_j\varepsilon\bigr]}_{\text{часть }(s)} \;-\; \nu - \mu R^2.
$$

Здесь $P$ -- матрица $n\times m$ со столбцами $p_j$.

---

### Двойственная функция

Двойственная функция

$$
g(\nu,\alpha,\lambda,\mu) = \inf_{x,\;s_j\ge\varepsilon} L(x,s,\ldots).
$$

Поскольку $L$ представлена в виде суммы функции от $x$ и суммы независимых функций от $s_j$ (плюс константы), инфимум равен сумме инфимумов.

#### Инфимум по $x$

Минимизируем $h(x) = \mu\|x\|_2^2 + c^\top x$ с $c = P\alpha + \nu\mathbf{1}$.

- Если $\mu > 0$, минимум достигается при $x^* = -c/(2\mu)$ и равен $-\|c\|^2/(4\mu)$.
- Если $\mu = 0$, инфимум конечен только при $c=0$ (иначе $-\infty$), и тогда равен $0$.

#### Инфимум по $s_j$ (для каждого $j$)

Для каждого $j$ решаем

$$
\inf_{s_j\ge\varepsilon} \bigl[-\pi_j\ln s_j - \beta_j s_j\bigr] + \lambda_j\varepsilon,\qquad \beta_j = \alpha_j+\lambda_j.
$$

При малом $\varepsilon$ оптимальное решение прямой задачи даёт $p_j^\top x > \varepsilon$, то есть ограничение неактивно. Тогда по дополняющей нежёсткости $\lambda_j = 0$ и $\beta_j = \alpha_j$.  
Для конечности инфимума необходимо $\alpha_j < 0$ (при $\alpha_j\ge 0$ линейный член $- \alpha_j s_j$ неограниченно убывает). При $\alpha_j < 0$ безусловный минимум достигается в точке $s_j^* = \pi_j/(-\alpha_j)$ и равен

$$
-\pi_j\ln\frac{\pi_j}{-\alpha_j} - \pi_j.
$$

(Случай попадания на границу $s_j=\varepsilon$ не реализуется в оптимуме, если $\varepsilon$ достаточно мало; поэтому в итоговой двойственной задаче $\lambda$ отсутствует.)

---

### Исключение $\alpha$ (замена $w$)

Введём $w_j := -\alpha_j > 0$. Тогда

- $\beta_j = -w_j < 0$,
- $c = P\alpha + \nu\mathbf{1} = -P w + \nu\mathbf{1}$.

Инфимум по $s$ даёт $\sum_j \bigl[-\pi_j\ln w_j + \pi_j\ln\pi_j - \pi_j\bigr]$.

Константы в $L$: $-\nu - \mu R^2$, плюс от $s$-части $+\pi_j\ln\pi_j$.

Собираем двойственную функцию:

$$
g(\nu,w,\mu) = \sum_{j=1}^m \pi_j \ln\frac{w_j}{\pi_j} - \sum_{j=1}^m \pi_j - \nu - \mu R^2 - \frac{\|P w - \nu\mathbf{1}\|_2^2}{4\mu}.
$$

Так как $\sum \pi_j = 1$, слагаемое $-\sum\pi_j$ превращается в $-1$.

---

### Двойственная задача

Двойственная задача -- максимизация $g$ по $w_j > 0$, $\nu\in\mathbb{R}$, $\mu\ge 0$. При $\mu=0$ эта формула доопределяется предельно:

$$
\boxed{\begin{aligned}
\text{max}_{w\ge 0,\,\nu,\,\mu\ge 0}\quad
&\sum_{j=1}^m \pi_j \ln\frac{w_j}{\pi_j} \;-\; 1 \;-\; \nu \;-\; \mu R^2 \;-\; \frac{\|P w - \nu\mathbf{1}\|_2^2}{4\mu},\\
\text{где при } \mu=0\ &\frac{\|P w - \nu\mathbf{1}\|^2}{4\mu} \triangleq \begin{cases}
0, & \text{если } P w = \nu\mathbf{1},\\
+\infty, & \text{иначе}.
\end{cases}
\end{aligned}}
\tag{D}
$$

 В коде для численной устойчивости вместо `4*mu` используется `4*mu + eps`, что соответствует добавлению малого положительного числа в знаменатель и не меняет сути задачи при стремлении `eps → 0`.

```python
import numpy as np
import cvxpy as cp
import pandas as pd
import time
from abc import ABC, abstractmethod
from typing import Dict, Any, List, Tuple

class OptimizationProblem(ABC):
    @abstractmethod
    def f(self, x: np.ndarray) -> float:
        pass

    @abstractmethod
    def grad(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def hess(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def constraints(self):
        pass

class Optimizer(ABC):
    @abstractmethod
    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        pass

    @abstractmethod
    def __str__(self) -> str:
        pass

class DataGenerator(ABC):
    @abstractmethod
    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        pass

    @abstractmethod
    def __str__(self) -> str:
        pass
```

```python
class BoundedLogOptimalProblem(OptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray, R: float = 100.0, eps: float = 1e-8):
        self.P = P
        self.pi = pi
        self.R = R
        self.eps = eps
        self.n = P.shape[0]
        self.m = P.shape[1]

    def f(self, x: np.ndarray) -> float:
        return -np.sum(self.pi * np.log(self.P.T @ x + 1e-16))

    def grad(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        return -self.P @ (self.pi / denom)

    def hess(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        tmp = self.pi / (denom ** 2)
        return self.P @ np.diag(tmp) @ self.P.T

    def constraints(self):
        A = np.ones((1, self.n))
        b = np.array([1.0])
        return A, b

class BoundedLogOptimalDualProblem(OptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray, R: float = 100.0):
        self.P = P
        self.pi = pi
        self.R = R
        self.n_var = P.shape[1] + 2
        self.m = P.shape[1]
        self.n_orig = P.shape[0]

    def _unpack(self, z: np.ndarray):
        w = z[:self.m]
        nu = z[self.m]
        mu = z[self.m + 1]
        return w, nu, mu

    def f(self, z: np.ndarray) -> float:
        w, nu, mu = self._unpack(z)
        obj = np.sum(self.pi * np.log(w / self.pi + 1e-16)) + 1 - nu - mu * self.R**2
        diff = self.P @ w - nu * np.ones(self.n_orig)
        obj -= np.dot(diff, diff) / (4 * mu + 1e-16)
        return -obj

    def grad(self, z: np.ndarray) -> np.ndarray:
        w, nu, mu = self._unpack(z)
        mu_safe = mu + 1e-16
        diff = self.P @ w - nu * np.ones(self.n_orig)
        Pdiff = self.P.T @ diff
        grad_w = -self.pi / (w + 1e-16) - Pdiff / (2 * mu_safe)
        sum_diff = np.sum(diff)
        grad_nu = -1.0 + sum_diff / (2 * mu_safe)
        grad_mu = self.R**2 + np.dot(diff, diff) / (4 * mu_safe ** 2)
        grad = np.concatenate([grad_w, [grad_nu], [grad_mu]])
        return grad

    def hess(self, z: np.ndarray) -> np.ndarray:
        w, nu, mu = self._unpack(z)
        mu_safe = mu + 1e-16
        diff = self.P @ w - nu * np.ones(self.n_orig)
        m = self.m
        H = np.zeros((m + 2, m + 2))
        H[:m, :m] = np.diag(self.pi / (w ** 2 + 1e-16)) + self.P.T @ self.P / (2 * mu_safe)
        Pdiff = self.P.T @ diff
        H[:m, m] = self.P.T @ np.ones(self.n_orig) / (2 * mu_safe)
        H[m, :m] = H[:m, m]
        H[:m, m+1] = Pdiff / (2 * mu_safe ** 2)
        H[m+1, :m] = H[:m, m+1]
        H[m, m] = -self.n_orig / (2 * mu_safe)
        sum_diff = np.sum(diff)
        H[m, m+1] = -sum_diff / (2 * mu_safe ** 2)
        H[m+1, m] = H[m, m+1]
        H[m+1, m+1] = -np.dot(diff, diff) / (2 * mu_safe ** 3)
        return H

    def constraints(self):
        return np.empty((0, self.n_var)), np.empty(0)
```

```python
class NormalDataGenerator(DataGenerator):
    def __init__(self, scale: float = 1.0, shift: float = 3.0):
        self.scale = scale
        self.shift = shift

    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        m = n * 2
        P = np.exp(np.random.randn(n, m) * self.scale + self.shift)
        P += np.eye(n, m) * n
        pi = np.random.rand(m)
        pi /= pi.sum()
        return P, pi

    def __str__(self) -> str:
        return f"Normal(scale={self.scale}, shift={self.shift})"

class UniformDataGenerator(DataGenerator):
    def __init__(self, low: float = 2.0, high: float = 7.0):
        self.low = low
        self.high = high

    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        m = n * 2
        P = np.exp(np.random.uniform(self.low, self.high, size=(n, m)))
        P += np.eye(n, m) * n
        pi = np.random.rand(m)
        pi /= pi.sum()
        return P, pi

    def __str__(self) -> str:
        return f"Uniform(low={self.low}, high={self.high})"
```

```python
class CVXPYBoundedOptimizer(Optimizer):
    def __init__(self, solver: str = 'CLARABEL'):
        self.solver = solver

    def __str__(self) -> str:
        return f"CVXPY({self.solver})"

    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        if isinstance(problem, BoundedLogOptimalProblem):
            return self._solve_primal(problem)
        elif isinstance(problem, BoundedLogOptimalDualProblem):
            return self._solve_dual(problem)
        else:
            raise TypeError

    def _solve_primal(self, prob: BoundedLogOptimalProblem) -> Dict[str, Any]:
        x = cp.Variable(prob.n)
        objective = cp.Minimize(-prob.pi @ cp.log(prob.P.T @ x))
        constraints = [
            cp.sum(x) == 1,
            prob.P.T @ x >= prob.eps,
            cp.norm(x) <= prob.R
        ]
        problem = cp.Problem(objective, constraints)
        t_start = time.time()
        opt_val = problem.solve(solver=self.solver)
        t_end = time.time()
        stats = problem.solver_stats
        return {
            'opt_val': opt_val,
            'opt_var': x.value,
            'time': t_end - t_start,
            'iters': stats.num_iters if stats else None,
            'status': problem.status
        }

    def _solve_dual(self, prob: BoundedLogOptimalDualProblem) -> Dict[str, Any]:
        w = cp.Variable(prob.m, nonneg=True)
        nu = cp.Variable()
        mu = cp.Variable(nonneg=True)
        tiny = 1e-12
        objective = cp.Maximize(
            prob.pi @ (cp.log(w / prob.pi + tiny)) + 1 - nu - mu * prob.R**2
            - cp.quad_over_lin(prob.P @ w - nu * np.ones(prob.n_orig), 4 * mu + tiny)
        )
        problem = cp.Problem(objective)
        t_start = time.time()
        opt_val = problem.solve(solver=self.solver)
        t_end = time.time()
        stats = problem.solver_stats
        return {
            'opt_val': opt_val,
            'opt_var': np.concatenate([w.value, [nu.value], [mu.value]]),
            'time': t_end - t_start,
            'iters': stats.num_iters if stats else None,
            'status': problem.status
        }
```

```python
class ExperimentRunnerBounded:
    def __init__(self,
                 n_list: List[int],
                 cases_per_n: int,
                 optimizers: List[Optimizer],
                 generators: List[DataGenerator],
                 R: float = 100.0,
                 eps: float = 1e-8):
        self.n_list = n_list
        self.cases_per_n = cases_per_n
        self.optimizers = optimizers
        self.generators = generators
        self.R = R
        self.eps = eps

    def run(self) -> pd.DataFrame:
        rows = []
        for n in self.n_list:
            for gen in self.generators:
                for case in range(self.cases_per_n):
                    P, pi = gen.generate(n)
                    primal = BoundedLogOptimalProblem(P, pi, self.R, self.eps)
                    dual = BoundedLogOptimalDualProblem(P, pi, self.R)
                    for opt in self.optimizers:
                        res_primal = opt.solve(primal)
                        res_dual = opt.solve(dual)
                        p_opt = res_primal['opt_val']
                        d_opt = res_dual['opt_val']
                        gap = abs(p_opt - d_opt)
                        rows.append({
                            'n': n,
                            'case': case,
                            'generator': str(gen),
                            'solver': str(opt),
                            'p_opt': p_opt,
                            'd_opt': d_opt,
                            'gap': gap,
                            'time_primal': res_primal['time'],
                            'iters_primal': res_primal['iters'],
                            'time_dual': res_dual['time'],
                            'iters_dual': res_dual['iters'],
                            'status_primal': res_primal['status'],
                            'status_dual': res_dual['status']
                        })
        return pd.DataFrame(rows)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

generators = [
    NormalDataGenerator(scale=1.0, shift=3.0),
    UniformDataGenerator(low=2.0, high=7.0)
]
optimizers = [CVXPYBoundedOptimizer(solver='CLARABEL')]

radii = [1, 10, 100]
df_list = []
for R_val in radii:
    runner = ExperimentRunnerBounded(
        n_list=list(range(10, 81, 10)),
        cases_per_n=5,
        optimizers=optimizers,
        generators=generators,
        R=R_val,
        eps=1e-8
    )
    df_r = runner.run()
    df_r['R'] = R_val
    df_list.append(df_r)
df_all = pd.concat(df_list, ignore_index=True)
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 120

def plot_performance_radius(df, solver_name=None):
    if solver_name:
        df = df[df['solver'] == solver_name].copy()
    else:
        solver_name = df['solver'].iloc[0]

    radii_order = sorted(df['R'].unique())
    n_radii = len(radii_order)

    agg = df.groupby(['n', 'generator', 'R']).agg(
        mean_gap=('gap', 'mean'),
        std_gap=('gap', 'std'),
        n_gap=('gap', 'count'),
        mean_time_primal=('time_primal', 'mean'),
        std_time_primal=('time_primal', 'std'),
        n_time_primal=('time_primal', 'count'),
        mean_iters_primal=('iters_primal', 'mean'),
        std_iters_primal=('iters_primal', 'std'),
        n_iters_primal=('iters_primal', 'count'),
        mean_time_dual=('time_dual', 'mean'),
        std_time_dual=('time_dual', 'std'),
        n_time_dual=('time_dual', 'count'),
        mean_iters_dual=('iters_dual', 'mean'),
        std_iters_dual=('iters_dual', 'std'),
        n_iters_dual=('iters_dual', 'count')
    ).reset_index()

    agg['ci_gap'] = 1.96 * agg['std_gap'] / np.sqrt(agg['n_gap'])
    agg['ci_time_primal'] = 1.96 * agg['std_time_primal'] / np.sqrt(agg['n_time_primal'])
    agg['ci_iters_primal'] = 1.96 * agg['std_iters_primal'] / np.sqrt(agg['n_iters_primal'])
    agg['ci_time_dual'] = 1.96 * agg['std_time_dual'] / np.sqrt(agg['n_time_dual'])
    agg['ci_iters_dual'] = 1.96 * agg['std_iters_dual'] / np.sqrt(agg['n_iters_dual'])

    generators = agg['generator'].unique()
    palette = sns.color_palette("Set1", len(generators))
    color_dict = dict(zip(generators, palette))

    fig, axes = plt.subplots(5, n_radii, figsize=(5 * n_radii, 16), sharex=True, sharey='row')
    if n_radii == 1:
        axes = axes.reshape(5, -1)

    for col_idx, R_val in enumerate(radii_order):
        data_r = agg[agg['R'] == R_val]
        ax_gap = axes[0, col_idx]
        ax_time_p = axes[1, col_idx]
        ax_iters_p = axes[2, col_idx]
        ax_time_d = axes[3, col_idx]
        ax_iters_d = axes[4, col_idx]

        for gen in generators:
            gen_data = data_r[data_r['generator'] == gen]
            x = gen_data['n']
            c = color_dict[gen]

            ax_gap.semilogy(x, gen_data['mean_gap'], 'o-', color=c, label=gen)
            ax_gap.fill_between(x,
                                np.maximum(1e-20, gen_data['mean_gap'] - gen_data['ci_gap']),
                                gen_data['mean_gap'] + gen_data['ci_gap'],
                                alpha=0.2, color=c)

            ax_time_p.plot(x, gen_data['mean_time_primal'], 's-', color=c)
            ax_time_p.fill_between(x,
                                   gen_data['mean_time_primal'] - gen_data['ci_time_primal'],
                                   gen_data['mean_time_primal'] + gen_data['ci_time_primal'],
                                   alpha=0.2, color=c)

            ax_iters_p.plot(x, gen_data['mean_iters_primal'], 'D-', color=c)
            ax_iters_p.fill_between(x,
                                    gen_data['mean_iters_primal'] - gen_data['ci_iters_primal'],
                                    gen_data['mean_iters_primal'] + gen_data['ci_iters_primal'],
                                    alpha=0.2, color=c)

            ax_time_d.plot(x, gen_data['mean_time_dual'], 'v-', color=c)
            ax_time_d.fill_between(x,
                                   gen_data['mean_time_dual'] - gen_data['ci_time_dual'],
                                   gen_data['mean_time_dual'] + gen_data['ci_time_dual'],
                                   alpha=0.2, color=c)

            ax_iters_d.plot(x, gen_data['mean_iters_dual'], 'p-', color=c)
            ax_iters_d.fill_between(x,
                                    gen_data['mean_iters_dual'] - gen_data['ci_iters_dual'],
                                    gen_data['mean_iters_dual'] + gen_data['ci_iters_dual'],
                                    alpha=0.2, color=c)

        ax_gap.set_title(f'R = {R_val}')
        ax_gap.grid(True, linestyle=':', alpha=0.7)
        ax_time_p.grid(True, linestyle=':', alpha=0.7)
        ax_iters_p.grid(True, linestyle=':', alpha=0.7)
        ax_time_d.grid(True, linestyle=':', alpha=0.7)
        ax_iters_d.grid(True, linestyle=':', alpha=0.7)

    axes[0, 0].set_ylabel('Средний gap')
    axes[1, 0].set_ylabel('Среднее время (primal), с')
    axes[2, 0].set_ylabel('Среднее число итераций (primal)')
    axes[3, 0].set_ylabel('Среднее время (dual), с')
    axes[4, 0].set_ylabel('Среднее число итераций (dual)')

    for col_idx in range(n_radii):
        axes[4, col_idx].set_xlabel('Размерность n')

    handles, labels = axes[0, 0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper center', ncol=len(generators), title='Генератор')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

plot_performance_radius(df_all, solver_name='CVXPY(CLARABEL)')
```

**Прямая задача**  
$$
\min_{x}\;\Bigl\{-\sum_{j=1}^{m}\pi_j\log(p_j^\top x)\;:\;\mathbf{1}^\top x=1,\;p_j^\top x>0\Bigr\},
$$  
где $\pi_j>0,\;\sum_{j}\pi_j=1$, $P=[p_1\;\dots\;p_m]\in\mathbb{R}^{n\times m}$. Введём переменную $t=P^\top x$ и запишем эквивалентную задачу  
$$
\min_{x,t}\;\Bigl\{-\sum_{j=1}^{m}\pi_j\log t_j\;:\;t=P^\top x,\;\mathbf{1}^\top x=1,\;t>0\Bigr\}.
$$

**Лагранжиан**  
Вводим множители $w\in\mathbb{R}^m$ для $t=P^\top x$ и $\nu\in\mathbb{R}$ для $\mathbf{1}^\top x=1$ (знаки выбраны так, чтобы двойственная функция была удобной):  
$$
\mathcal{L}(x,t,w,\nu)=-\sum_{j=1}^{m}\pi_j\log t_j+\sum_{j=1}^{m}w_j(t_j-p_j^\top x)+\nu(\mathbf{1}^\top x-1).
$$

**Минимизация по $x$**  
$$
\nabla_x\mathcal{L}= -P w+\nu\mathbf{1}=0 \quad\Longrightarrow\quad P w=\nu\mathbf{1}.
$$  
Если это условие не выполнено, $\inf_x\mathcal{L}=-\infty$.

**Минимизация по $t_j>0$**  
$$
\frac{\partial\mathcal{L}}{\partial t_j}=-\frac{\pi_j}{t_j}+w_j=0 \quad\Longrightarrow\quad t_j=\frac{\pi_j}{w_j},\quad w_j>0.
$$  
Подстановка даёт  
$$
-\pi_j\log\frac{\pi_j}{w_j}+w_j\frac{\pi_j}{w_j}= \pi_j\log\frac{w_j}{\pi_j}+\pi_j.
$$

**Двойственная функция**  
При ограничениях $P w=\nu\mathbf{1},\;w>0$  
$$
\begin{aligned}
g(w,\nu)&=\inf_{x,t>0}\mathcal{L}= \sum_{j=1}^{m}\Bigl(\pi_j\log\frac{w_j}{\pi_j}+\pi_j\Bigr)-\nu \\
&= \sum_{j=1}^{m}\pi_j\log\frac{w_j}{\pi_j}+1-\nu,
\end{aligned}
$$  
так как $\sum\pi_j=1$.

**Двойственная задача**  
$$
\boxed{\max_{w,\nu}\;\sum_{j=1}^{m}\pi_j\log\frac{w_j}{\pi_j}+1-\nu,\qquad P w=\nu\mathbf{1},\;w>0.}
$$  
Слабая двойственность выполняется по построению. Сильная двойственность имеет место (например, при $p_j>0$ задача удовлетворяет условию Слейтера), поэтому оптимальные значения совпадают.

```python
class LogOptimalProblem(OptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray):
        self.P = P
        self.pi = pi
        self.n = P.shape[0]
        self.m = P.shape[1]

    def f(self, x: np.ndarray) -> float:
        return -np.sum(self.pi * np.log(self.P.T @ x + 1e-16))

    def grad(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        return -self.P @ (self.pi / denom)

    def hess(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        tmp = self.pi / (denom ** 2)
        return self.P @ np.diag(tmp) @ self.P.T

    def constraints(self):
        A = np.ones((1, self.n))
        b = np.array([1.0])
        return A, b

class LogOptimalDualProblem(OptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray):
        self.P = P
        self.pi = pi
        self.n_orig, self.m = P.shape
        self.n_var = self.m + 1

    def _unpack(self, z: np.ndarray):
        w = z[:self.m]
        nu = z[self.m]
        return w, nu

    def f(self, z: np.ndarray) -> float:
        w, nu = self._unpack(z)
        obj = np.sum(self.pi * np.log(w / self.pi + 1e-16)) + 1 - nu
        return -obj

    def grad(self, z: np.ndarray) -> np.ndarray:
        w, nu = self._unpack(z)
        grad_w = -self.pi / (w + 1e-16)
        grad_nu = 1.0
        return np.concatenate([grad_w, [grad_nu]])

    def hess(self, z: np.ndarray) -> np.ndarray:
        w, nu = self._unpack(z)
        H = np.zeros((self.n_var, self.n_var))
        H[:self.m, :self.m] = np.diag(self.pi / (w ** 2 + 1e-16))
        return H

    def constraints(self):
        A = np.zeros((self.n_orig, self.n_var))
        A[:, :self.m] = self.P
        A[:, self.m] = -1.0
        b = np.zeros(self.n_orig)
        return A, b
```

```python
class CVXPYOptimizer(Optimizer):
    def __init__(self, solver: str = 'CLARABEL'):
        self.solver = solver

    def __str__(self) -> str:
        return f"CVXPY({self.solver})"

    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        if isinstance(problem, LogOptimalProblem):
            return self._solve_primal(problem)
        elif isinstance(problem, LogOptimalDualProblem):
            return self._solve_dual(problem)
        else:
            raise TypeError

    def _solve_primal(self, prob: LogOptimalProblem) -> Dict[str, Any]:
        x = cp.Variable(prob.n)
        objective = cp.Minimize(-prob.pi @ cp.log(prob.P.T @ x))
        constraints = [
            cp.sum(x) == 1
        ]
        problem = cp.Problem(objective, constraints)
        t_start = time.time()
        opt_val = problem.solve(solver=self.solver)
        t_end = time.time()
        stats = problem.solver_stats
        return {
            'opt_val': opt_val,
            'opt_var': x.value,
            'time': t_end - t_start,
            'iters': stats.num_iters if stats else None,
            'status': problem.status
        }

    def _solve_dual(self, prob: LogOptimalDualProblem) -> Dict[str, Any]:
        w = cp.Variable(prob.m, nonneg=True)
        nu = cp.Variable()
        objective = cp.Maximize(
            prob.pi @ cp.log(w / prob.pi + 1e-12) + 1 - nu
        )
        constraints = [
            prob.P @ w == nu * np.ones(prob.n_orig)
        ]
        problem = cp.Problem(objective, constraints)
        t_start = time.time()
        opt_val = problem.solve(solver=self.solver)
        t_end = time.time()
        stats = problem.solver_stats
        return {
            'opt_val': opt_val,
            'opt_var': np.concatenate([w.value, [nu.value]]),
            'time': t_end - t_start,
            'iters': stats.num_iters if stats else None,
            'status': problem.status
        }
```

```python
class DNMOptimizer(Optimizer):
    def __init__(self, tol: float = 1e-8, max_iter: int = 1000, alpha: float = 0.25, beta: float = 0.5):
        self.tol = tol
        self.max_iter = max_iter
        self.alpha = alpha
        self.beta = beta
        self.is_dual = False

    def __str__(self) -> str:
        return "DNM"

    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        A, b = problem.constraints()

        n = getattr(problem, 'n_var', getattr(problem, 'n', None))

        if hasattr(problem, 'pi') and hasattr(problem, 'n_var'):
            P = problem.P
            n_assets, m_scenarios = P.shape

            A_start = np.concatenate([P, -np.ones((n_assets, 1))], axis=1)

            target_w = problem.pi
            target_nu = np.mean(P @ target_w)
            z_target = np.concatenate([target_w, [target_nu]])

            v, _, _, _ = np.linalg.lstsq(A_start, A_start @ z_target, rcond=None)
            x_feasible = z_target - v

            w_part = x_feasible[:m_scenarios]
            if np.any(w_part <= 0):
                w_part = np.maximum(w_part, 1e-4)
                x_feasible[:m_scenarios] = w_part

            x = x_feasible
            self.is_dual = True
        elif hasattr(problem, 'n'):
            x = np.ones(n) / n
            self.is_dual = False

        t_start = time.time()
        status = "optimal"
        iters = 0
        final_f = problem.f(x)

        for i in range(self.max_iter):
            iters = i
            f = problem.f(x)
            grad = problem.grad(x)
            hess = problem.hess(x)

            if A.shape[0] > 0:
                kkt_matrix = np.block([
                    [hess, A.T],
                    [A, np.zeros((A.shape[0], A.shape[0]))]
                ])
                rhs = np.concatenate([-grad, np.zeros(A.shape[0])])
                try:
                    sol = np.linalg.solve(kkt_matrix, rhs)
                    step = sol[:n]
                except np.linalg.LinAlgError:
                    status = "singular_hessian"; break
            else:
                try:
                    step = np.linalg.solve(hess, -grad)
                except np.linalg.LinAlgError:
                    status = "singular_hessian"; break

            f_decrement_sq = -grad.dot(step)
            if f_decrement_sq < 0:
                break
            if f_decrement_sq / 2.0 <= self.tol:
                final_f = f
                break

            t = 1.0
            while True:
                x_next = x + t * step
                f_next = problem.f(x_next)

                if not np.isnan(f_next) and f_next <= f + self.alpha * t * grad.dot(step):
                    break

                t *= self.beta
                if t < 1e-20:
                    status = "line_search_failed"
                    break

            if status != "optimal":
                break

            x = x_next
            final_f = f_next

        t_end = time.time()

        return {
            'opt_val': -final_f if self.is_dual else final_f,
            'opt_var': x,
            'time': t_end - t_start,
            'iters': iters,
            'status': status
        }
```

# Комбинированная схема

```python
import numpy as np
import time

def accelerated_gradient_method(f_grad, x0, A, b, L0=1.0, eta=1.5,
                                max_iter=500, tol=1e-8, restart=True,
                                nonneg_indices=None):
    m, n = A.shape
 
    if not np.allclose(A @ x0, b):
        x0 = x0 - A.T @ np.linalg.solve(A @ A.T, A @ x0 - b)
    if nonneg_indices is not None:
        x0[nonneg_indices] = np.maximum(x0[nonneg_indices], 1e-12)

    AAT = A @ A.T

    def proj(z):
        y = np.linalg.solve(AAT, A @ z - b)
        out = z - A.T @ y
        if nonneg_indices is not None:
            out[nonneg_indices] = np.maximum(out[nonneg_indices], 1e-12)
        return out

    def proj_tangent(g):
        y = np.linalg.solve(AAT, A @ g)
        return g - A.T @ y

    x = x0.copy()
    y = x0.copy()
    t = 1.0
    L = L0

    for k in range(1, max_iter + 1):
        f_y, g_y = f_grad(y)
        if not np.isfinite(f_y):
            L = L0
            y = x.copy()
            t = 1.0
            continue

        L_try = L
        while True:
            x_new = proj(y - (1.0 / L_try) * g_y)
            f_new, _ = f_grad(x_new)
            if np.isfinite(f_new):
                diff = x_new - y
                Q = f_y + np.dot(g_y, diff) + 0.5 * L_try * np.dot(diff, diff)
                if f_new <= Q + 1e-12:
                    break
            L_try *= eta

        L = max(L_try / eta, 1e-12)

        x_prev = x.copy()
        x = x_new

        if restart and f_new > f_grad(x_prev)[0] + 1e-12:
            t = 1.0
            y = x.copy()
        else:
            t_prev = t
            t = 0.5 * (1.0 + np.sqrt(1.0 + 4.0 * t_prev * t_prev))
            y = x + ((t_prev - 1.0) / t) * (x - x_prev)

        _, g = f_grad(x)
        if np.linalg.norm(proj_tangent(g)) < tol:
            return x, -f_grad(x)[0], k, None
    return x, -f_grad(x)[0], max_iter, None


class AcceleratedGradientOptimizer(Optimizer):
    def __init__(self, L0=1.0, eta=1.5, max_iter=500, tol=1e-8, restart=True):
        self.L0 = L0
        self.eta = eta
        self.max_iter = max_iter
        self.tol = tol
        self.restart = restart

    def __str__(self):
        return "AcceleratedGradient"

    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        A, b = problem.constraints()
        n = getattr(problem, 'n_var', getattr(problem, 'n', None))
        nonneg_idx = None

        if isinstance(problem, LogOptimalDualProblem):
            w0 = np.ones(problem.m) / problem.m
            nu0 = np.mean(problem.P @ w0)
            x0 = np.concatenate([w0, [nu0]])
            nonneg_idx = np.arange(problem.m)  
        elif isinstance(problem, LogOptimalProblem):
            x0 = np.ones(problem.n) / problem.n
        else:
            x0 = np.zeros(n)

        def f_grad(x):
            f = problem.f(x)
            g = problem.grad(x)
            return f, g

        t_start = time.time()
        x_opt, f_opt, iters, _ = accelerated_gradient_method(
            f_grad, x0, A, b,
            L0=self.L0, eta=self.eta,
            max_iter=self.max_iter, tol=self.tol,
            restart=self.restart,
            nonneg_indices=nonneg_idx
        )
        t_end = time.time()

        opt_val = -problem.f(x_opt) if isinstance(problem, LogOptimalDualProblem) else problem.f(x_opt)

        return {
            'opt_val': opt_val,
            'opt_var': x_opt,
            'time': t_end - t_start,
            'iters': iters,
            'status': 'optimal' if np.isfinite(opt_val) else 'failed'
        }
```

```python
def combined_method(P, pi, x0=None, A=None, b=None,
                    grad_tol=1e-6, newton_tol=1e-12,
                    grad_max_iter=10000, newton_max_iter=100,
                    alpha=0.25, beta=0.5):
    n = P.shape[0]
    if x0 is None:
        x0 = np.ones(n) / n
    if A is None:
        A = np.ones((1, n))
        b = np.ones(1)

    f_grad = lambda x: log_opt_f_grad(x, P, pi)
    x_agm, _, iters_agm, _ = accelerated_gradient_method(
        f_grad, x0, A, b,
        max_iter=grad_max_iter, tol=grad_tol, restart=True
    )
    f_agm, _ = log_opt_f_grad(x_agm, P, pi)


    prob = LogOptimalProblem(P, pi)
    x = x_agm.copy()
    status = "optimal"
    iters = 0
    final_f = f_agm

    for i in range(newton_max_iter):
        iters = i
        f = prob.f(x)
        grad = prob.grad(x)
        hess = prob.hess(x)

        kkt_matrix = np.block([
            [hess, A.T],
            [A, np.zeros((A.shape[0], A.shape[0]))]
        ])
        rhs = np.concatenate([-grad, np.zeros(A.shape[0])])
        try:
            sol = np.linalg.solve(kkt_matrix, rhs)
            step = sol[:n]
        except np.linalg.LinAlgError:
            status = "singular_hessian"
            break

        f_decrement_sq = -grad.dot(step)
        if f_decrement_sq < 0:
            break
        if f_decrement_sq / 2.0 <= newton_tol:
            final_f = f
            break

        t = 1.0
        while True:
            x_next = x + t * step
            f_next = prob.f(x_next)
            if not np.isnan(f_next) and f_next <= f + alpha * t * grad.dot(step):
                break
            t *= beta
            if t < 1e-20:
                status = "line_search_failed"
                break
        if status != "optimal":
            break
        x = x_next
        final_f = f_next

    f_min = final_f
    f_max = -f_min

    summary = {
        'x_agm': x_agm,
        'f_agm_min': f_agm,
        'iters_agm': iters_agm,
        'x_opt': x,
        'f_opt': f_max,
        'f_min': f_min,
        'iters_newton': iters,
        'iters_total': iters_agm + iters,
        'nu_opt': None
    }
    return x, f_max, summary


class AccelNewtonOptimizer(Optimizer):
    def __init__(self, grad_tol=1e-6, newton_tol=1e-12,
                 grad_max_iter=10000, newton_max_iter=100,
                 alpha=0.25, beta=0.5):
        self.grad_tol = grad_tol
        self.newton_tol = newton_tol
        self.grad_max_iter = grad_max_iter
        self.newton_max_iter = newton_max_iter
        self.alpha = alpha
        self.beta = beta

    def __str__(self):
        return "AccelNewton"

    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        if isinstance(problem, LogOptimalProblem):
            P, pi = problem.P, problem.pi
            A, b = problem.constraints()
            start_time = time.time()
            x_opt, f_max, summary = combined_method(
                P, pi,
                A=A, b=b,
                grad_tol=self.grad_tol,
                newton_tol=self.newton_tol,
                grad_max_iter=self.grad_max_iter,
                newton_max_iter=self.newton_max_iter,
                alpha=self.alpha,
                beta=self.beta
            )
            total_time = time.time() - start_time
            return {
                'opt_val': -f_max,        # в формате прямой задачи
                'opt_var': x_opt,
                'time': total_time,
                'iters': summary['iters_total'],
                'status': 'optimal'
            }
        elif isinstance(problem, LogOptimalDualProblem):
            dnm = DNMOptimizer()
            return dnm.solve(problem)
        else:
            raise TypeError("Поддерживаются только LogOptimalProblem и LogOptimalDualProblem")
```

# SR1 и DFP 

```python

from scipy.linalg import null_space
from scipy.optimize import linprog

def normalize_probabilities(pi):
    pi = np.asarray(pi, dtype=float).reshape(-1)
    if np.any(pi <= 0):
        raise ValueError("Все компоненты pi должны быть положительными.")
    return pi / pi.sum()


def generate_simple_log_opt_data(n, delta=0.35, seed=None):
    rng = np.random.default_rng(seed)
    eye = np.eye(n)
    ones_col = np.ones((n, 1))
    P = np.hstack([ones_col + delta * eye, ones_col - delta * eye])

    pi = rng.uniform(0.5, 1.5, size=2 * n)
    pi = normalize_probabilities(pi)
    return P, pi


def primal_value_grad_simple(x, P, pi, tiny=1e-14):
    x = np.asarray(x, dtype=float).reshape(-1)
    P = np.asarray(P, dtype=float)
    pi = normalize_probabilities(pi)

    s = P.T @ x
    if np.any(s <= tiny) or not np.all(np.isfinite(s)):
        return np.inf, np.zeros_like(x)

    f = -float(np.dot(pi, np.log(s)))
    grad = -(P @ (pi / s))
    return f, grad


def dual_value_grad_simple(z, P, pi, tiny=1e-14):
    z = np.asarray(z, dtype=float).reshape(-1)
    P = np.asarray(P, dtype=float)
    pi = normalize_probabilities(pi)

    m = P.shape[1]
    w = z[:m]
    nu = z[m]

    if np.any(w <= tiny) or not np.isfinite(nu):
        return np.inf, np.zeros_like(z)

    f = -float(np.dot(pi, np.log(w / pi))) - 1.0 + float(nu)
    grad = np.concatenate([-pi / w, [1.0]])
    return f, grad


def project_to_affine(x, A, b):
    x = np.asarray(x, dtype=float).reshape(-1)
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)

    if A.size == 0 or A.shape[0] == 0:
        return x

    return x - A.T @ np.linalg.pinv(A @ A.T) @ (A @ x - b)


def find_dual_feasible_start_simple(P, tiny=1e-8):
    P = np.asarray(P, dtype=float)
    n, m = P.shape

    c = np.zeros(m + 2)
    c[-1] = -1.0

    A_eq_main = np.hstack([P, -np.ones((n, 1)), np.zeros((n, 1))])
    A_eq_sum = np.zeros((1, m + 2))
    A_eq_sum[0, :m] = 1.0
    A_eq = np.vstack([A_eq_main, A_eq_sum])
    b_eq = np.concatenate([np.zeros(n), [1.0]])

    A_ub = np.zeros((m, m + 2))
    A_ub[:, :m] = -np.eye(m)
    A_ub[:, -1] = 1.0
    b_ub = np.zeros(m)

    bounds = [(0.0, None)] * m + [(None, None), (0.0, None)]

    res = linprog(
        c,
        A_ub=A_ub,
        b_ub=b_ub,
        A_eq=A_eq,
        b_eq=b_eq,
        bounds=bounds,
        method="highs",
    )

    if not res.success or res.x[-1] <= tiny:
        raise ValueError(
            "Не удалось найти строго допустимую стартовую точку для двойственной задачи. "
            "Проверьте генератор P или используйте generate_simple_log_opt_data(...)."
        )

    z0 = res.x[:m + 1]
    A_dual = np.hstack([P, -np.ones((n, 1))])
    if np.linalg.norm(A_dual @ z0) > 1e-7 or np.min(z0[:m]) <= tiny:
        raise ValueError("Стартовая точка двойственной задачи получилась недопустимой.")

    return z0


def quasi_newton_affine(
    value_grad,
    x0,
    A,
    b,
    method="SR1",
    f_star=None,
    gap_tol=1e-2,
    grad_tol=1e-8,
    max_iter=3000,
    alpha=1e-4,
    beta=0.5,
):
    method = method.upper()
    if method not in {"SR1", "DFP"}:
        raise ValueError("method должен быть 'SR1' или 'DFP'.")

    x0 = np.asarray(x0, dtype=float).reshape(-1)
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)

    x_base = project_to_affine(x0, A, b)
    Z = null_space(A)
    r = Z.shape[1]

    if r == 0:
        f, g = value_grad(x_base)
        return {
            "x": x_base,
            "f": f,
            "iterations": 0,
            "status": "zero_dimensional",
            "time": 0.0,
            "history": [{"iter": 0, "f": f, "gap": None, "grad_norm": 0.0, "step": 0.0}],
        }

    u = np.zeros(r)
    H = np.eye(r)

    x = x_base.copy()
    f, g = value_grad(x)

    history = []
    status = "max_iter_reached"
    start_time = time.time()

    for k in range(max_iter + 1):
        grad_reduced = Z.T @ g
        grad_norm = float(np.linalg.norm(grad_reduced))
        gap = None if f_star is None else abs(float(f) - float(f_star))

        history.append({
            "iter": k,
            "f": float(f),
            "gap": gap,
            "grad_norm": grad_norm,
            "step": 0.0,
        })

        if np.isfinite(f) and ((gap is not None and gap <= gap_tol) or (gap is None and grad_norm <= grad_tol)):
            status = "optimal"
            break

        if k == max_iter:
            break

        direction = -H @ grad_reduced

        if not np.all(np.isfinite(direction)) or grad_reduced @ direction >= -1e-14:
            H = np.eye(r)
            direction = -grad_reduced

        directional_derivative = float(grad_reduced @ direction)
        step = 1.0
        accepted = False

        for _ in range(100):
            u_next = u + step * direction
            x_next = x_base + Z @ u_next
            f_next, g_next = value_grad(x_next)

            if np.isfinite(f_next) and f_next <= f + alpha * step * directional_derivative:
                accepted = True
                break

            step *= beta

        history[-1]["step"] = float(step)

        if not accepted:
            status = "line_search_failed"
            break

        grad_reduced_next = Z.T @ g_next
        s = step * direction
        y = grad_reduced_next - grad_reduced

        if method == "SR1":
            v = s - H @ y
            denom = float(v @ y)

            if abs(denom) > 1e-10 * np.linalg.norm(v) * max(1.0, np.linalg.norm(y)):
                H = H + np.outer(v, v) / denom

        elif method == "DFP":
            sy = float(s @ y)
            Hy = H @ y
            yHy = float(y @ Hy)

            if sy > 1e-12 and yHy > 1e-12:
                H = H + np.outer(s, s) / sy - np.outer(Hy, Hy) / yHy

        H = 0.5 * (H + H.T)

        if not np.all(np.isfinite(H)) or np.linalg.norm(H, ord="fro") > 1e10:
            H = np.eye(r)

        u = u_next
        x = x_next
        f = f_next
        g = g_next

    return {
        "x": x,
        "f": float(f),
        "iterations": len(history) - 1,
        "status": status,
        "time": time.time() - start_time,
        "history": history,
    }


def solve_reference_simple(P, pi, solver="CLARABEL"):
    P = np.asarray(P, dtype=float)
    pi = normalize_probabilities(pi)
    n, m = P.shape

    x = cp.Variable(n)
    primal_problem = cp.Problem(
        cp.Minimize(-pi @ cp.log(P.T @ x)),
        [cp.sum(x) == 1],
    )

    t0 = time.time()
    p_opt = primal_problem.solve(solver=solver)
    p_time = time.time() - t0

    w = cp.Variable(m)
    nu = cp.Variable()
    tiny = 1e-12
    dual_problem = cp.Problem(
        cp.Maximize(pi @ cp.log(w / pi + tiny) + 1 - nu),
        [P @ w == nu * np.ones(n), w >= tiny],
    )

    t0 = time.time()
    d_opt = dual_problem.solve(solver=solver)
    d_time = time.time() - t0

    if x.value is None or w.value is None or nu.value is None:
        raise RuntimeError(
            f"CVXPY не вернул решение: primal_status={primal_problem.status}, "
            f"dual_status={dual_problem.status}"
        )

    return {
        "primal": {
            "opt_val": float(p_opt),
            "opt_var": np.asarray(x.value).reshape(-1),
            "status": primal_problem.status,
            "iters": primal_problem.solver_stats.num_iters,
            "time": p_time,
            "source": f"CVXPY({solver})",
        },
        "dual": {
            "opt_val": float(d_opt),
            "opt_var": np.concatenate([np.asarray(w.value).reshape(-1), [float(nu.value)]]),
            "status": dual_problem.status,
            "iters": dual_problem.solver_stats.num_iters,
            "time": d_time,
            "source": f"CVXPY({solver})",
        },
    }


def run_sr1_dfp_comparison(n=10, seed=0, delta=0.35, gap_tol=1e-2, max_iter=3000):
    P, pi = generate_simple_log_opt_data(n=n, delta=delta, seed=seed)
    reference = solve_reference_simple(P, pi)

    p_star = reference["primal"]["opt_val"]
    d_star = reference["dual"]["opt_val"]

    A_primal = np.ones((1, n))
    b_primal = np.array([1.0])
    x0 = np.ones(n) / n

    A_dual = np.hstack([P, -np.ones((n, 1))])
    b_dual = np.zeros(n)
    z0 = find_dual_feasible_start_simple(P)

    rows = [{
        "method": reference["primal"]["source"],
        "p_value": p_star,
        "d_value": d_star,
        "primal_dual_gap": abs(p_star - d_star),
        "gap_p_to_ref": 0.0,
        "gap_d_to_ref": 0.0,
        "iters_primal": reference["primal"]["iters"],
        "iters_dual": reference["dual"]["iters"],
        "time_primal": reference["primal"]["time"],
        "time_dual": reference["dual"]["time"],
        "status_primal": reference["primal"]["status"],
        "status_dual": reference["dual"]["status"],
        "eq_err_primal": np.linalg.norm(A_primal @ reference["primal"]["opt_var"] - b_primal),
        "eq_err_dual": np.linalg.norm(A_dual @ reference["dual"]["opt_var"]),
        "min_PTx": np.min(P.T @ reference["primal"]["opt_var"]),
        "min_w": np.min(reference["dual"]["opt_var"][:P.shape[1]]),
    }]

    histories = {}

    for method in ["SR1", "DFP"]:
        primal_res = quasi_newton_affine(
            lambda x: primal_value_grad_simple(x, P, pi),
            x0,
            A_primal,
            b_primal,
            method=method,
            f_star=p_star,
            gap_tol=gap_tol,
            max_iter=max_iter,
        )

        dual_res_min = quasi_newton_affine(
            lambda z: dual_value_grad_simple(z, P, pi),
            z0,
            A_dual,
            b_dual,
            method=method,
            f_star=-d_star,
            gap_tol=gap_tol,
            max_iter=max_iter,
        )

        p_value = primal_res["f"]
        d_value = -dual_res_min["f"]

        rows.append({
            "method": method,
            "p_value": p_value,
            "d_value": d_value,
            "primal_dual_gap": abs(p_value - d_value),
            "gap_p_to_ref": abs(p_value - p_star),
            "gap_d_to_ref": abs(d_value - d_star),
            "iters_primal": primal_res["iterations"],
            "iters_dual": dual_res_min["iterations"],
            "time_primal": primal_res["time"],
            "time_dual": dual_res_min["time"],
            "status_primal": primal_res["status"],
            "status_dual": dual_res_min["status"],
            "eq_err_primal": np.linalg.norm(A_primal @ primal_res["x"] - b_primal),
            "eq_err_dual": np.linalg.norm(A_dual @ dual_res_min["x"]),
            "min_PTx": np.min(P.T @ primal_res["x"]),
            "min_w": np.min(dual_res_min["x"][:P.shape[1]]),
        })

        histories[(method, "primal")] = primal_res["history"]
        histories[(method, "dual")] = dual_res_min["history"]

    comparison = pd.DataFrame(rows)

    checks = comparison[comparison["method"].isin(["SR1", "DFP"])]
    assert np.all(checks["gap_p_to_ref"] <= gap_tol + 1e-8)
    assert np.all(checks["gap_d_to_ref"] <= gap_tol + 1e-8)
    assert np.all(checks["eq_err_primal"] <= 1e-7)
    assert np.all(checks["eq_err_dual"] <= 1e-7)
    assert np.all(checks["min_PTx"] > 0)
    assert np.all(checks["min_w"] > 0)

    return comparison, histories, P, pi


def run_sr1_dfp_benchmark(n_values=(10, 20, 30), cases_per_n=3, gap_tol=1e-2, max_iter=3000):
    rows = []

    for n in n_values:
        for case in range(cases_per_n):
            comparison, _, _, _ = run_sr1_dfp_comparison(
                n=n,
                seed=1000 + 100 * n + case,
                gap_tol=gap_tol,
                max_iter=max_iter,
            )

            for _, row in comparison[comparison["method"].isin(["SR1", "DFP"])].iterrows():
                rows.append({
                    "n": n,
                    "case": case,
                    "method": row["method"],
                    "p_value": row["p_value"],
                    "d_value": row["d_value"],
                    "primal_dual_gap": row["primal_dual_gap"],
                    "gap_p_to_ref": row["gap_p_to_ref"],
                    "gap_d_to_ref": row["gap_d_to_ref"],
                    "iters_primal": row["iters_primal"],
                    "iters_dual": row["iters_dual"],
                    "time_primal": row["time_primal"],
                    "time_dual": row["time_dual"],
                    "status_primal": row["status_primal"],
                    "status_dual": row["status_dual"],
                })

    raw = pd.DataFrame(rows)

    summary = (
        raw
        .groupby(["n", "method"], as_index=False)
        .agg({
            "iters_primal": "mean",
            "iters_dual": "mean",
            "time_primal": "mean",
            "time_dual": "mean",
            "primal_dual_gap": "mean",
            "gap_p_to_ref": "mean",
            "gap_d_to_ref": "mean",
        })
    )

    return raw, summary
```

### Прямая задача (минимизация)

$$
\begin{aligned}
\min_{x\in\mathbb{R}^n}\quad & -\sum_{j=1}^m \pi_j \ln(p_j^\top x) \\
\text{s.t.}\quad & \mathbf{1}^\top x = 1, \\
& p_j^\top x \ge \varepsilon > 0,\quad j=1,\dots,m,\\
& \|x\|_2 \le R.
\end{aligned}
\tag{P}
$$

**Выпуклость**  
- $-\ln t$ выпукла на $(0,\infty)$;  
- композиция $-\ln(p_j^\top x)$ с аффинной $p_j^\top x$ выпукла на $\{x\mid p_j^\top x>0\}$;  
- на $\mathcal{F}$ имеем $p_j^\top x\ge\varepsilon>0$, поэтому каждое слагаемое $-\pi_j\ln(p_j^\top x)$ выпукло;  
- сумма выпуклых функций с неотрицательными весами $\pi_j$ выпукла $\Rightarrow$ $f$ выпукла;  
- $\mathcal{F}$ -- пересечение аффинного подпространства, полупространств и шара -- замкнуто, выпукло и компактно.

**Условие Слейтера**  
Предположим, что существует портфель $\bar{x}$ с $\mathbf{1}^\top\bar{x}=1$ и $p_j^\top\bar{x}>0$ для всех $j$.
Выбрав $\varepsilon\in(0,\;\min_j p_j^\top\bar{x})$ и $R>\|\bar{x}\|_2$, получим строго допустимую точку:
$$
\mathbf{1}^\top\bar{x}=1,\quad p_j^\top\bar{x}>\varepsilon\;\forall j,\quad \|\bar{x}\|_2<R,
$$
то есть условие Слейтера выполнено.

**Сильная двойственность**  
Задача выпукла, $\mathcal{F}$ непусто и компактно, условие Слейтера выполнено $\Longrightarrow$ сильная двойственность: оптимальные значения прямой и двойственной задач совпадают.

---

### Введение вспомогательных переменных

Введём $s_j = p_j^\top x$. Задача (P) эквивалентна

$$
\begin{aligned}
\min_{x,\,s}\quad & -\sum_{j=1}^m \pi_j \ln s_j \\
\text{s.t.}\quad & s_j = p_j^\top x, \\
& \mathbf{1}^\top x = 1, \\
& s_j \ge \varepsilon, \\
& \|x\|_2 \le R.
\end{aligned}
\tag{1}
$$

Целевая функция зависит только от $s$, ограничения связывают $x$ и $s$.

---

### Лагранжиан

Двойственные переменные:
- $\nu\in\mathbb{R}$ для $\mathbf{1}^\top x = 1$,
- $\alpha_j\in\mathbb{R}$ для $s_j = p_j^\top x$,
- $\lambda_j\ge 0$ для $s_j \ge \varepsilon$,
- $\mu\ge 0$ для $\|x\|_2^2 \le R^2$.

Лагранжиан (ограничения‑неравенства приведены к виду $g(x)\le 0$):

$$
\begin{aligned}
L(x,s,\nu,\alpha,\lambda,\mu) = &-\sum_{j}\pi_j\ln s_j \\
&+ \nu(\mathbf{1}^\top x - 1) \\
&+ \sum_{j}\alpha_j(p_j^\top x - s_j) \\
&+ \sum_{j}\lambda_j(\varepsilon - s_j) \\
&+ \mu\bigl(\|x\|_2^2 - R^2\bigr).
\end{aligned}
$$

Перегруппируем слагаемые, выделяя части, зависящие только от $x$ и только от $s$:

$$
L = \underbrace{\bigl[\mu\|x\|_2^2 + (P\alpha + \nu\mathbf{1})^\top x\bigr]}_{\text{часть }(x)} \;+\; \underbrace{\sum_{j}\bigl[-\pi_j\ln s_j - (\alpha_j+\lambda_j)s_j + \lambda_j\varepsilon\bigr]}_{\text{часть }(s)} \;-\; \nu - \mu R^2.
$$

Здесь $P$ -- матрица $n\times m$ со столбцами $p_j$.

---

### Двойственная функция

Двойственная функция

$$
g(\nu,\alpha,\lambda,\mu) = \inf_{x,\;s_j\ge\varepsilon} L(x,s,\ldots).
$$

Поскольку $L$ представлена в виде суммы функции от $x$ и суммы независимых функций от $s_j$ (плюс константы), инфимум равен сумме инфимумов.

#### Инфимум по $x$

Минимизируем $h(x) = \mu\|x\|_2^2 + c^\top x$ с $c = P\alpha + \nu\mathbf{1}$.

- Если $\mu > 0$, минимум достигается при $x^* = -c/(2\mu)$ и равен $-\|c\|^2/(4\mu)$.
- Если $\mu = 0$, инфимум конечен только при $c=0$ (иначе $-\infty$), и тогда равен $0$.

#### Инфимум по $s_j$ (для каждого $j$)

Для каждого $j$ решаем

$$
\inf_{s_j\ge\varepsilon} \bigl[-\pi_j\ln s_j - \beta_j s_j\bigr] + \lambda_j\varepsilon,\qquad \beta_j = \alpha_j+\lambda_j.
$$

При малом $\varepsilon$ оптимальное решение прямой задачи даёт $p_j^\top x > \varepsilon$, то есть ограничение неактивно. Тогда по дополняющей нежёсткости $\lambda_j = 0$ и $\beta_j = \alpha_j$.  
Для конечности инфимума необходимо $\alpha_j < 0$ (при $\alpha_j\ge 0$ линейный член $- \alpha_j s_j$ неограниченно убывает). При $\alpha_j < 0$ безусловный минимум достигается в точке $s_j^* = \pi_j/(-\alpha_j)$ и равен

$$
-\pi_j\ln\frac{\pi_j}{-\alpha_j} - \pi_j.
$$

(Случай попадания на границу $s_j=\varepsilon$ не реализуется в оптимуме, если $\varepsilon$ достаточно мало; поэтому в итоговой двойственной задаче $\lambda$ отсутствует.)

---

### Исключение $\alpha$ (замена $w$)

Введём $w_j := -\alpha_j > 0$. Тогда

- $\beta_j = -w_j < 0$,
- $c = P\alpha + \nu\mathbf{1} = -P w + \nu\mathbf{1}$.

Инфимум по $s$ даёт $\sum_j \bigl[-\pi_j\ln w_j + \pi_j\ln\pi_j - \pi_j\bigr]$.

Константы в $L$: $-\nu - \mu R^2$, плюс от $s$-части $+\pi_j\ln\pi_j$.

Собираем двойственную функцию:

$$
g(\nu,w,\mu) = \sum_{j=1}^m \pi_j \ln\frac{w_j}{\pi_j} - \sum_{j=1}^m \pi_j - \nu - \mu R^2 - \frac{\|P w - \nu\mathbf{1}\|_2^2}{4\mu}.
$$

Так как $\sum \pi_j = 1$, слагаемое $-\sum\pi_j$ превращается в $-1$.

---

### Двойственная задача

Двойственная задача -- максимизация $g$ по $w_j > 0$, $\nu\in\mathbb{R}$, $\mu\ge 0$. При $\mu=0$ эта формула доопределяется предельно:

$$
\boxed{\begin{aligned}
\text{max}_{w\ge 0,\,\nu,\,\mu\ge 0}\quad 
&\sum_{j=1}^m \pi_j \ln\frac{w_j}{\pi_j} \;-\; 1 \;-\; \nu \;-\; \mu R^2 \;-\; \frac{\|P w - \nu\mathbf{1}\|_2^2}{4\mu},\\
\text{где при } \mu=0\ &\frac{\|P w - \nu\mathbf{1}\|^2}{4\mu} \triangleq \begin{cases}
0, & \text{если } P w = \nu\mathbf{1},\\
+\infty, & \text{иначе}.
\end{cases}
\end{aligned}}
\tag{D}
$$

 В коде для численной устойчивости вместо `4*mu` используется `4*mu + eps`, что соответствует добавлению малого положительного числа в знаменатель и не меняет сути задачи при стремлении `eps → 0`.

```python
import numpy as np
import cvxpy as cp
import pandas as pd
import time
from abc import ABC, abstractmethod
from typing import Dict, Any, List, Tuple

class OptimizationProblem(ABC):
    @abstractmethod
    def f(self, x: np.ndarray) -> float:
        pass

    @abstractmethod
    def grad(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def hess(self, x: np.ndarray) -> np.ndarray:
        pass

    @abstractmethod
    def constraints(self):
        pass

class Optimizer(ABC):
    @abstractmethod
    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        pass

    @abstractmethod
    def __str__(self) -> str:
        pass

class DataGenerator(ABC):
    @abstractmethod
    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        pass

    @abstractmethod
    def __str__(self) -> str:
        pass
```

```python
class LogOptimalProblem(OptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray, R: float = 100.0, eps: float = 1e-8):
        self.P = P
        self.pi = pi
        self.R = R
        self.eps = eps
        self.n = P.shape[0]
        self.m = P.shape[1]

    def f(self, x: np.ndarray) -> float:
        return -np.sum(self.pi * np.log(self.P.T @ x + 1e-16))

    def grad(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        return -self.P @ (self.pi / denom)

    def hess(self, x: np.ndarray) -> np.ndarray:
        denom = self.P.T @ x + 1e-16
        tmp = self.pi / (denom ** 2)
        return self.P @ np.diag(tmp) @ self.P.T

    def constraints(self):
        A = np.ones((1, self.n))
        b = np.array([1.0])
        return A, b

class LogOptimalDualProblem(OptimizationProblem):
    def __init__(self, P: np.ndarray, pi: np.ndarray, R: float = 100.0):
        self.P = P
        self.pi = pi
        self.R = R
        self.n_var = P.shape[1] + 2
        self.m = P.shape[1]
        self.n_orig = P.shape[0]

    def _unpack(self, z: np.ndarray):
        w = z[:self.m]
        nu = z[self.m]
        mu = z[self.m + 1]
        return w, nu, mu

    def f(self, z: np.ndarray) -> float:
        w, nu, mu = self._unpack(z)
        obj = np.sum(self.pi * np.log(w / self.pi + 1e-16)) + 1 - nu - mu * self.R**2
        diff = self.P @ w - nu * np.ones(self.n_orig)
        obj -= np.dot(diff, diff) / (4 * mu + 1e-16)
        return -obj

    def grad(self, z: np.ndarray) -> np.ndarray:
        w, nu, mu = self._unpack(z)
        mu_safe = mu + 1e-16
        diff = self.P @ w - nu * np.ones(self.n_orig)
        Pdiff = self.P.T @ diff
        grad_w = -self.pi / (w + 1e-16) - Pdiff / (2 * mu_safe)
        sum_diff = np.sum(diff)
        grad_nu = -1.0 + sum_diff / (2 * mu_safe)
        grad_mu = -self.R**2 + np.dot(diff, diff) / (4 * mu_safe ** 2)
        grad = np.concatenate([grad_w, [grad_nu], [grad_mu]])
        return grad

    def hess(self, z: np.ndarray) -> np.ndarray:
        w, nu, mu = self._unpack(z)
        mu_safe = mu + 1e-16
        diff = self.P @ w - nu * np.ones(self.n_orig)
        m = self.m
        H = np.zeros((m + 2, m + 2))
        H[:m, :m] = np.diag(self.pi / (w ** 2 + 1e-16)) + self.P.T @ self.P / (2 * mu_safe)
        Pdiff = self.P.T @ diff
        H[:m, m] = self.P.T @ np.ones(self.n_orig) / (2 * mu_safe)
        H[m, :m] = H[:m, m]
        H[:m, m+1] = Pdiff / (2 * mu_safe ** 2)
        H[m+1, :m] = H[:m, m+1]
        H[m, m] = -self.n_orig / (2 * mu_safe)
        sum_diff = np.sum(diff)
        H[m, m+1] = -sum_diff / (2 * mu_safe ** 2)
        H[m+1, m] = H[m, m+1]
        H[m+1, m+1] = -np.dot(diff, diff) / (2 * mu_safe ** 3)
        return H

    def constraints(self):
        return np.empty((0, self.n_var)), np.empty(0)
```

```python
class NormalDataGenerator(DataGenerator):
    def __init__(self, scale: float = 2.0, shift: float = 1.0):
        self.scale = scale
        self.shift = shift

    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        m = n * 2
        P = np.random.randn(n, m) * self.scale + self.shift
        pi = np.random.rand(m)
        pi /= pi.sum()
        return P, pi

    def __str__(self) -> str:
        return f"Normal(scale={self.scale}, shift={self.shift})"

class UniformDataGenerator(DataGenerator):
    def __init__(self, low: float = -2.0, high: float = 3.0):
        self.low = low
        self.high = high

    def generate(self, n: int) -> Tuple[np.ndarray, np.ndarray]:
        m = n * 2
        P = np.random.uniform(self.low, self.high, size=(n, m))
        pi = np.random.rand(m)
        pi /= pi.sum()
        return P, pi

    def __str__(self) -> str:
        return f"Uniform(low={self.low}, high={self.high})"
```

```python
class CVXPYOptimizer(Optimizer):
    def __init__(self, solver: str = 'CLARABEL'):
        self.solver = solver

    def __str__(self) -> str:
        return f"CVXPY({self.solver})"

    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        if isinstance(problem, LogOptimalProblem):
            return self._solve_primal(problem)
        elif isinstance(problem, LogOptimalDualProblem):
            return self._solve_dual(problem)
        else:
            raise TypeError

    def _solve_primal(self, prob: LogOptimalProblem) -> Dict[str, Any]:
        x = cp.Variable(prob.n)
        objective = cp.Minimize(-prob.pi @ cp.log(prob.P.T @ x))
        constraints = [
            cp.sum(x) == 1,
            prob.P.T @ x >= prob.eps,
            cp.norm(x) <= prob.R
        ]
        problem = cp.Problem(objective, constraints)
        t_start = time.time()
        opt_val = problem.solve(solver=self.solver)
        t_end = time.time()
        stats = problem.solver_stats
        return {
            'opt_val': opt_val,
            'opt_var': x.value,
            'time': t_end - t_start,
            'iters': stats.num_iters if stats else None,
            'status': problem.status
        }

    def _solve_dual(self, prob: LogOptimalDualProblem) -> Dict[str, Any]:
        w = cp.Variable(prob.m, nonneg=True)
        nu = cp.Variable()
        mu = cp.Variable(nonneg=True)
        tiny = 1e-12
        objective = cp.Maximize(
            prob.pi @ (cp.log(w / prob.pi + tiny)) + 1 - nu - mu * prob.R**2
            - cp.quad_over_lin(prob.P @ w - nu * np.ones(prob.n_orig), 4 * mu + tiny)
        )
        problem = cp.Problem(objective)
        t_start = time.time()
        opt_val = problem.solve(solver=self.solver)
        t_end = time.time()
        stats = problem.solver_stats
        return {
            'opt_val': opt_val,
            'opt_var': np.concatenate([w.value, [nu.value], [mu.value]]),
            'time': t_end - t_start,
            'iters': stats.num_iters if stats else None,
            'status': problem.status
        }
```

```python
class ExperimentRunner:
    def __init__(self,
                 n_list: List[int],
                 cases_per_n: int,
                 optimizers: List[Optimizer],
                 generators: List[DataGenerator],
                 R: float = 100.0,
                 eps: float = 1e-8):
        self.n_list = n_list
        self.cases_per_n = cases_per_n
        self.optimizers = optimizers
        self.generators = generators
        self.R = R
        self.eps = eps

    def run(self) -> pd.DataFrame:
        rows = []
        for n in self.n_list:
            for gen in self.generators:
                for case in range(self.cases_per_n):
                    P, pi = gen.generate(n)
                    primal = LogOptimalProblem(P, pi, self.R, self.eps)
                    dual = LogOptimalDualProblem(P, pi, self.R)
                    for opt in self.optimizers:
                        res_primal = opt.solve(primal)
                        res_dual = opt.solve(dual)
                        p_opt = res_primal['opt_val']
                        d_opt = res_dual['opt_val']
                        gap = abs(p_opt - d_opt)
                        rows.append({
                            'n': n,
                            'case': case,
                            'generator': str(gen),
                            'solver': str(opt),
                            'p_opt': p_opt,
                            'd_opt': d_opt,
                            'gap': gap,
                            'time_primal': res_primal['time'],
                            'iters_primal': res_primal['iters'],
                            'time_dual': res_dual['time'],
                            'iters_dual': res_dual['iters'],
                            'status_primal': res_primal['status'],
                            'status_dual': res_dual['status']
                        })
        return pd.DataFrame(rows)

if __name__ == "__main__":
    np.random.seed(42)

    generators = [
        NormalDataGenerator(scale=2.0, shift=1.0),
        UniformDataGenerator(low=-2.0, high=3.0)
    ]

    optimizers = [
        CVXPYOptimizer(solver='CLARABEL')
    ]

    runner = ExperimentRunner(
        n_list=[i for i in range(10, 21, 10)],
        cases_per_n=3,
        optimizers=optimizers,
        generators=generators,
        R=100.0,
        eps=1e-8
    )

    df = runner.run()
    print(df)
    df.to_csv("experiment_results.csv", index=False)
```

```python
from __future__ import annotations

import time
from dataclasses import dataclass
from typing import Callable, Dict, Any, Optional, Tuple

import numpy as np


Array = np.ndarray


def normalize_P(P: Array, pi: Array) -> Array:
    P = np.asarray(P, dtype=float)
    pi = np.asarray(pi, dtype=float).reshape(-1)
    if P.ndim != 2:
        raise ValueError("P must be a matrix")
    if P.shape[0] == pi.size:
        return P
    if P.shape[1] == pi.size:
        return P.T
    raise ValueError(
        f"Cannot match P shape {P.shape} with len(pi)={pi.size}. "
        "Expected P.shape[0] == len(pi) or P.shape[1] == len(pi)."
    )


def log_opt_value_grad(P: Array, pi: Array, eps: float = 1e-8, R: float = 100.0):
    P = normalize_P(P, pi)
    pi = np.asarray(pi, dtype=float).reshape(-1)
    pi = pi / pi.sum()

    def oracle(x: Array) -> Tuple[float, Array]:
        x = np.asarray(x, dtype=float).reshape(-1)
        ret = P @ x
        if np.any(ret < eps) or np.linalg.norm(x) > R * (1.0 + 1e-10):
            return np.inf, np.zeros_like(x)
        f = -float(np.dot(pi, np.log(ret)))
        grad = -(P.T @ (pi / ret))
        return f, grad

    return oracle


def equality_nullspace(A: Array, tol: float = 1e-12) -> Array:
    A = np.asarray(A, dtype=float)
    if A.size == 0:
        n = A.shape[1] if A.ndim == 2 else 0
        return np.eye(n)
    _, s, vt = np.linalg.svd(A, full_matrices=True)
    if s.size == 0:
        rank = 0
    else:
        rank = int(np.sum(s > tol * max(A.shape) * s[0]))
    return vt[rank:].T.copy()


def project_to_equality(x: Array, A: Array, b: Array) -> Array:
    x = np.asarray(x, dtype=float).reshape(-1)
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)
    if A.size == 0:
        return x.copy()
    residual = A @ x - b
    if np.linalg.norm(residual) < 1e-12:
        return x.copy()
    correction = A.T @ np.linalg.solve(A @ A.T, residual)
    return x - correction


def project_ball(z: Array, radius: Optional[float]) -> Array:
    if radius is None or not np.isfinite(radius):
        return z.copy()
    nz = np.linalg.norm(z)
    if nz <= radius:
        return z.copy()
    if nz == 0:
        return z.copy()
    return z * (radius / nz)


@dataclass
class QuasiNewtonResult:
    x: Array
    f: float
    iterations: int
    status: str
    method: str
    time: float
    history: Dict[str, list]

    def as_dict(self) -> Dict[str, Any]:
        return {
            "x": self.x,
            "f": self.f,
            "iterations": self.iterations,
            "status": self.status,
            "method": self.method,
            "time": self.time,
            "history": self.history,
        }


def quasi_newton_equality(
    f_grad: Callable[[Array], Tuple[float, Array]],
    x0: Array,
    A: Optional[Array] = None,
    b: Optional[Array] = None,
    *,
    method: str = "SR1",
    R: Optional[float] = None,
    f_star: Optional[float] = None,
    accuracy: float = 1e-2,
    grad_tol: float = 1e-8,
    max_iter: int = 2000,
    alpha: float = 0.25,
    beta: float = 0.5,
    min_step: float = 1e-14,
    update_tol: float = 1e-10,
    H0_scale: float = 1.0,
) -> Dict[str, Any]:

    method = method.upper()
    if method not in {"SR1", "DFP"}:
        raise ValueError("method must be 'SR1' or 'DFP'")

    x0 = np.asarray(x0, dtype=float).reshape(-1)
    n = x0.size
    if A is None:
        A = np.empty((0, n))
    else:
        A = np.asarray(A, dtype=float)
    if b is None:
        b = np.empty(0)
    else:
        b = np.asarray(b, dtype=float).reshape(-1)

    # Minimum-norm point in the affine set Ax=b.  With an orthonormal nullspace N,
    # every feasible point has x = x_aff + N z and ||x||^2 = ||x_aff||^2 + ||z||^2.
    x_aff = project_to_equality(np.zeros(n), A, b)
    N = equality_nullspace(A)
    r = N.shape[1]

    if R is None:
        z_radius = None
    else:
        residual_radius_sq = float(R * R - np.dot(x_aff, x_aff))
        if residual_radius_sq < -1e-10:
            raise ValueError("The equality set Ax=b does not intersect the ball ||x|| <= R.")
        z_radius = float(np.sqrt(max(0.0, residual_radius_sq)))

    if r == 0:
        f0, _ = f_grad(x_aff)
        return QuasiNewtonResult(
            x=x_aff,
            f=f0,
            iterations=0,
            status="no degrees of freedom",
            method=method,
            time=0.0,
            history={"f": [f0], "gap": [None], "pg_norm": [0.0], "step": [0.0], "updates_skipped": [False]},
        ).as_dict()

    x0_eq = project_to_equality(x0, A, b)
    z = N.T @ (x0_eq - x_aff)
    z = project_ball(z, z_radius)

    def x_from_z(z_: Array) -> Array:
        return x_aff + N @ z_

    def phi_grad(z_: Array) -> Tuple[float, Array]:
        x = x_from_z(z_)
        f, gx = f_grad(x)
        if not np.isfinite(f):
            return np.inf, np.zeros(r)
        return f, N.T @ gx

    def projected_gradient_norm(z_: Array, g_: Array) -> float:
        return float(np.linalg.norm(z_ - project_ball(z_ - g_, z_radius)))

    f, g = phi_grad(z)
    if not np.isfinite(f):
        raise ValueError(
            "Initial point is outside the domain; choose feasible x0 with P @ x0 >= eps and ||x0|| <= R."
        )

    H = H0_scale * np.eye(r)
    history = {
        "f": [float(f)],
        "gap": [None if f_star is None else abs(float(f) - f_star)],
        "pg_norm": [projected_gradient_norm(z, g)],
        "step": [0.0],
        "updates_skipped": [False],
    }

    status = "max_iter reached"
    t_start = time.perf_counter()

    for k in range(1, max_iter + 1):
        gap = None if f_star is None else abs(f - f_star)
        pg_norm = projected_gradient_norm(z, g)
        if (gap is not None and gap <= accuracy) or pg_norm <= grad_tol:
            status = "converged"
            k -= 1
            break

        H = 0.5 * (H + H.T)
        p = -H @ g
        if not np.all(np.isfinite(p)):
            H = H0_scale * np.eye(r)
            p = -g

        def try_direction(direction: Array):
            step = 1.0
            while step >= min_step:
                z_raw = z + step * direction
                z_trial = project_ball(z_raw, z_radius)
                s_trial = z_trial - z
                if np.linalg.norm(s_trial) <= 1e-15:
                    step *= beta
                    continue
                model_decrease = float(g @ s_trial)
                if model_decrease >= 0:
                    step *= beta
                    continue
                f_trial, g_trial = phi_grad(z_trial)
                if np.isfinite(f_trial) and f_trial <= f + alpha * model_decrease:
                    return True, step, z_trial, f_trial, g_trial
                step *= beta
            return False, 0.0, z, f, g

        accepted, step, z_trial, f_trial, g_trial = try_direction(p)
        if not accepted:
            H = H0_scale * np.eye(r)
            p = -g
            accepted, step, z_trial, f_trial, g_trial = try_direction(p)
            if not accepted:
                status = "line search failed"
                k -= 1
                break

        s = z_trial - z
        y = g_trial - g
        Hy = H @ y
        skipped = False

        if method == "SR1":
            u = s - Hy
            denom = float(u @ y)
            threshold = update_tol * np.linalg.norm(u) * np.linalg.norm(y)
            if abs(denom) > threshold:
                H = H + np.outer(u, u) / denom
            else:
                skipped = True
        else:  # DFP
            sy = float(s @ y)
            yHy = float(y @ Hy)
            threshold_sy = update_tol * np.linalg.norm(s) * np.linalg.norm(y)
            threshold_yHy = update_tol * np.linalg.norm(y) * max(1.0, np.linalg.norm(Hy))
            if sy > threshold_sy and yHy > threshold_yHy:
                H = H + np.outer(s, s) / sy - np.outer(Hy, Hy) / yHy
            else:
                skipped = True

        if not np.all(np.isfinite(H)):
            H = H0_scale * np.eye(r)
            skipped = True

        z, f, g = z_trial, f_trial, g_trial
        history["f"].append(float(f))
        history["gap"].append(None if f_star is None else abs(float(f) - f_star))
        history["pg_norm"].append(projected_gradient_norm(z, g))
        history["step"].append(float(step))
        history["updates_skipped"].append(skipped)
    else:
        k = max_iter

    elapsed = time.perf_counter() - t_start
    x = x_from_z(z)
    return QuasiNewtonResult(
        x=x,
        f=float(f),
        iterations=int(k),
        status=status,
        method=method,
        time=elapsed,
        history=history,
    ).as_dict()


def solve_primal_quasi_newton(
    P: Array,
    pi: Array,
    *,
    method: str = "SR1",
    x0: Optional[Array] = None,
    eps: float = 1e-8,
    R: float = 100.0,
    f_star: Optional[float] = None,
    accuracy: float = 1e-2,
    max_iter: int = 2000,
) -> Dict[str, Any]:
    P = normalize_P(P, pi)
    n = P.shape[1]
    if x0 is None:
        x0 = np.ones(n) / n
    A = np.ones((1, n))
    b = np.array([1.0])
    oracle = log_opt_value_grad(P, pi, eps=eps, R=R)
    return quasi_newton_equality(
        oracle,
        x0,
        A,
        b,
        method=method,
        R=R,
        f_star=f_star,
        accuracy=accuracy,
        max_iter=max_iter,
    )


def solve_primal_cvxpy(
    P: Array,
    pi: Array,
    eps: float = 1e-8,
    R: float = 100.0,
    solver: str = "CLARABEL",
) -> Dict[str, Any]:
    import cvxpy as cp

    P = normalize_P(P, pi)
    pi = np.asarray(pi, dtype=float).reshape(-1)
    pi = pi / pi.sum()
    _, n = P.shape
    x = cp.Variable(n)
    objective = cp.Minimize(-cp.sum(cp.multiply(pi, cp.log(P @ x))))
    constraints = [cp.sum(x) == 1, P @ x >= eps, cp.norm(x, 2) <= R]
    problem = cp.Problem(objective, constraints)
    t0 = time.perf_counter()
    try:
        value = problem.solve(solver=solver, verbose=False)
    except Exception:
        value = problem.solve(verbose=False)
    elapsed = time.perf_counter() - t0
    return {
        "x": None if x.value is None else np.asarray(x.value).reshape(-1),
        "f": float(value) if value is not None else np.nan,
        "status": problem.status,
        "time": elapsed,
        "iterations": None if problem.solver_stats is None else problem.solver_stats.num_iters,
    }


def make_positive_test_instance(n: int, m: Optional[int] = None, seed: int = 0) -> Tuple[Array, Array]:
    rng = np.random.default_rng(seed)
    if m is None:
        m = 2 * n
    P = rng.uniform(0.5, 2.0, size=(m, n))
    pi = rng.dirichlet(np.ones(m))
    return P, pi


def test_sr1_dfp_against_cvxpy(seed: int = 0) -> None:
    P, pi = make_positive_test_instance(n=10, m=20, seed=seed)
    ref = solve_primal_cvxpy(P, pi, eps=1e-8, R=100.0)
    if ref["status"] not in {"optimal", "optimal_inaccurate"}:
        raise AssertionError(f"CVXPY did not solve the test problem: {ref['status']}")

    for method in ("SR1", "DFP"):
        res = solve_primal_quasi_newton(
            P, pi, method=method, f_star=ref["f"], accuracy=1e-2, max_iter=2000
        )
        gap = abs(res["f"] - ref["f"])
        feasibility_sum = abs(np.sum(res["x"]) - 1.0)
        min_return = float(np.min(P @ res["x"]))
        print(
            f"{method}: status={res['status']}, iters={res['iterations']}, "
            f"f={res['f']:.10f}, cvx={ref['f']:.10f}, gap={gap:.2e}, "
            f"sum_gap={feasibility_sum:.2e}, min(Px)={min_return:.2e}, "
            f"norm={np.linalg.norm(res['x']):.4f}"
        )
        assert gap <= 1e-2 + 1e-8, f"{method} objective gap too large: {gap:.2e}"
        assert feasibility_sum <= 1e-10, f"{method} violates equality"
        assert min_return >= 1e-8 - 1e-10, f"{method} violates log-domain lower bound"
        assert np.linalg.norm(res["x"]) <= 100.0 + 1e-8, f"{method} violates ball constraint"


if __name__ == "__main__":
    test_sr1_dfp_against_cvxpy()
```

```python
class QuasiNewtonCOptimizer(Optimizer):
    def __init__(self, method: str = "SR1", accuracy: float = 1e-2, max_iter: int = 2000):
        method = method.upper()
        if method not in {"SR1", "DFP"}:
            raise ValueError("method must be 'SR1' or 'DFP'")
        self.method = method
        self.accuracy = accuracy
        self.max_iter = max_iter

    def __str__(self) -> str:
        return f"{self.method}(participant C)"

    def solve(self, problem: OptimizationProblem) -> Dict[str, Any]:
        if isinstance(problem, LogOptimalProblem):
            return self._solve_primal(problem)
        if isinstance(problem, LogOptimalDualProblem):
            return self._solve_dual(problem)
        raise TypeError(f"Unsupported problem type: {type(problem)}")

    def _solve_primal(self, prob: LogOptimalProblem) -> Dict[str, Any]:
        x0 = np.ones(prob.n) / prob.n
        res = solve_primal_quasi_newton(
            prob.P, prob.pi,
            method=self.method,
            x0=x0,
            eps=prob.eps,
            R=prob.R,
            f_star=None,
            accuracy=self.accuracy,
            max_iter=self.max_iter,
        )
        return {
            "opt_val": res["f"],
            "opt_var": res["x"],
            "time": res["time"],
            "iters": res["iterations"],
            "status": res["status"],
            "history": res["history"],
        }

    def _solve_dual(self, prob: LogOptimalDualProblem) -> Dict[str, Any]:
        m = prob.m

        def unpack(q):
            u = np.clip(q[:m], -50.0, 50.0)
            nu = q[m]
            v = np.clip(q[m + 1], -50.0, 50.0)
            w = np.exp(u)
            mu = np.exp(v)
            return w, nu, mu

        def dual_oracle(q):
            w, nu, mu = unpack(q)
            z = np.concatenate([w, [nu, mu]])
            f = prob.f(z)          # f = -dual_objective
            grad_z = prob.grad(z)
            if not np.isfinite(f) or not np.all(np.isfinite(grad_z)):
                return np.inf, np.zeros_like(q)
            grad_q = np.concatenate([grad_z[:m] * w, [grad_z[m]], [grad_z[m + 1] * mu]])
            return f, grad_q

        q0 = np.concatenate([
            np.log(np.maximum(prob.pi, 1e-8)),
            [0.0, 0.0]   # nu=0, mu=1
        ])
        res = quasi_newton_equality(
            dual_oracle,
            q0,
            method=self.method,
            f_star=None,
            accuracy=self.accuracy,
            max_iter=self.max_iter,
        )
        w, nu, mu = unpack(res["x"])
        z_opt = np.concatenate([w, [nu, mu]])
        return {
            "opt_val": -res["f"],
            "opt_var": z_opt,
            "time": res["time"],
            "iters": res["iterations"],
            "status": res["status"],
            "history": res["history"],
        }
```

```python
P_test, pi_test = make_positive_test_instance(n=10, m=20, seed=0)
ref = solve_primal_cvxpy(P_test, pi_test, eps=1e-8, R=100.0)
print("CVXPY:", ref["status"], "f*=", ref["f"])

for method in ["SR1", "DFP"]:
    res = solve_primal_quasi_newton(
        P_test, pi_test,
        method=method,
        f_star=ref["f"],
        accuracy=1e-2,
        max_iter=2000,
    )
    gap = abs(res["f"] - ref["f"])
    print(
        f"{method}: status={res['status']}, iters={res['iterations']}, "
        f"f={res['f']:.8f}, gap={gap:.2e}, "
        f"sum(x)-1={np.sum(res['x'])-1:.2e}, "
        f"min(Px)={np.min(P_test @ res['x']):.2e}, "
        f"||x||={np.linalg.norm(res['x']):.4f}"
    )
    assert gap <= 1e-2 + 1e-8
    assert abs(np.sum(res["x"]) - 1.0) <= 1e-10
    assert np.min(P_test @ res["x"]) >= 1e-8 - 1e-10
    assert np.linalg.norm(res["x"]) <= 100.0 + 1e-8
```
